# 烟雾与火焰报警系统

## 项目简介

本项目实现一个基于 Arduino Uno 的烟雾与火焰报警系统，联合使用 MQ-2 和火焰传感器判断环境风险，并通过蜂鸣器、状态指示灯、继电器以及 LCD1602 I2C 进行报警联动与状态显示。

系统使用双传感器联合判定方式，将风险分为正常、烟雾预警和火灾报警三级，并加入确认时间与恢复保持时间，避免偶发噪声导致误触发或频繁来回切换。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| MQ-2 | 1 | 烟雾或可燃气体检测 |
| 火焰传感器 | 1 | 火焰检测 |
| 有源蜂鸣器 | 1 | 声音报警 |
| 继电器模块 | 1 | 联动风扇、水泵或断电装置 |
| LED | 3 | 绿色正常、黄色预警、红色报警 |
| LCD1602 I2C | 1 | 显示传感器数值和系统状态 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 检测烟雾浓度变化。
- 检测火焰是否出现。
- 支持烟雾预警状态。
- 支持火灾报警状态。
- 蜂鸣器按不同节奏输出提示。
- 继电器在火灾报警时联动外部设备。
- LCD1602 实时显示烟雾数值、火焰状态和系统模式。
- 使用确认时间和恢复保持时间降低误报警风险。

## 运行方式

1. 使用 Arduino IDE 打开 `src/smoke_and_flame_alarm_system/smoke_and_flame_alarm_system.ino`。
2. 安装 `LiquidCrystal_I2C` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。


## 状态设计

| 状态 | 说明 |
| --- | --- |
| 正常 | 烟雾值正常且未检测到火焰，系统持续监测 |
| 烟雾预警 | MQ-2 超过预警阈值并持续一段时间，蜂鸣器慢速提示 |
| 火灾报警 | 检测到火焰或烟雾达到更高阈值，蜂鸣器快速报警并吸合继电器 |

- 进入预警或报警前需要满足确认时间，避免瞬时尖峰误触发。
- 条件恢复正常后不会立即解除，而是等待恢复保持时间结束，再回到正常状态。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 正常监测 | 绿色 LED 亮，LCD 显示正常状态 |
| 烟雾预警 | 提高 MQ-2 输出后进入预警状态，黄色 LED 亮，蜂鸣器慢速鸣叫 |
| 火焰报警 | 检测到火焰后进入报警状态，红色 LED 亮，继电器吸合 |
| 继电器联动 | 报警状态下继电器输出有效，可联动外部负载 |
| 状态保持 | 风险消失后系统不会瞬间跳回正常，而是经过保持时间恢复 |
| LCD 显示 | 可显示烟雾值、火焰状态和当前工作模式 |


## 可扩展方向

- 增加 GSM 或 WiFi 模块，实现远程报警。
- 增加 EEPROM 保存阈值配置。
- 增加风扇、水泵或电磁阀的更复杂联动策略。
- 增加事件记录功能，保存报警发生次数和时间。
- 增加多区域传感器输入，形成多点火灾监测系统。
