# 巡线小车

## 项目简介

本项目实现一个基于 Arduino、三路红外循迹模块和双电机驱动的小型巡线小车。系统通过左、中、右三个传感器判断黑线相对车体中心的位置，再根据偏差计算左右轮差速，实现基础直线跟随和弯道修正。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino 主控板 | 1 | 控制小车运动 |
| 三路红外循迹模块 | 1 组 | 检测黑线位置 |
| 双路电机驱动 | 1 | 驱动左右电机 |
| 直流减速电机 | 2 | 作为驱动轮 |
| 小车底盘与电源 | 1 套 | 承载与供电 |


## 核心功能

- 三路红外模拟量转在线 / 离线状态。
- 根据传感器组合推断黑线偏左、居中或偏右。
- 通过左右轮差速控制修正方向。
- 丢线时主动停车，避免盲目前冲。
- 串口输出各传感器状态和误差，便于调阈值。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 左循迹传感器 | A0 |
| 中循迹传感器 | A1 |
| 右循迹传感器 | A2 |
| 左电机 PWM / 方向 | D5、D6、D7 |
| 右电机 PWM / 方向 | D9、D10、D11 |


## 接线说明

- 三路循迹模块的电源接 5V 和 GND，模拟输出分别接 A0、A1、A2。
- 左右电机经驱动模块接入主控，PWM 引脚接使能端，方向引脚接 IN1/IN2。
- 传感器安装高度要尽量一致，否则同一条黑线会在三路上表现出不同阈值。
- 测试前建议先在白底与黑线处分别读取模拟值，重新调整 `LINE_THRESHOLD`。


## 运行方式

1. 打开 `src/line_following_car/line_following_car.ino`。
2. 完成循迹模块与电机驱动接线。
3. 依据实际赛道调整 `LINE_THRESHOLD` 和 `TURN_GAIN`。
4. 上传程序后在黑线轨道上测试直线和弯道表现。
5. 通过串口观察 `L/C/R` 状态和误差值。


## 控制逻辑说明

- `readLineState()` 会把模拟量转换为二值状态，低于阈值时认为压在线上。
- `calculateError()` 依据左右偏差返回 `-1`、`0`、`1`，表示应向左修、居中或向右修。
- 主循环以 `BASE_SPEED` 为基础速度，再通过 `TURN_GAIN * error` 形成左右轮速度差。
- 三路都离线时说明完全丢线，系统立即停车而不是继续按上一帧误差前进。
- 当前版本使用简单比例修正，适合基础巡线入门和后续升级到 PID。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 直线跟随 | 小车能沿黑线稳定前进 |
| 弯道修正 | 轻微弯道处能主动调节方向 |
| 状态输出 | 串口能看到三路状态与误差 |
| 丢线保护 | 完全离线时小车停车 |


## 可扩展方向

- 增加 PID 巡线，提高弯道和高速稳定性。
- 增加路口识别、岔路处理和停车点识别。
- 增加编码器反馈，形成速度闭环。
- 增加蓝牙手动 / 自动切换模式。
