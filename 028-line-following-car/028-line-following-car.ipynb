# 巡线小车

## 项目简介

本项目实现一个基于 Arduino、红外循迹模块和双电机驱动的小型巡线小车，使用三路传感器判断黑线相对位置，并通过调整左右电机速度实现直线跟随和弯道修正。


## 主要器件

Arduino、小车底盘、3 路循迹模块、双电机驱动、直流减速电机。


## 核心功能

- 三路循迹传感输入。
- 黑线位置判断。
- 左右电机差速控制。
- 丢线停车保护。


## 运行方式

1. 打开 `src/line_following_car/line_following_car.ino`。
2. 连接三路循迹模块和双电机驱动。
3. 调整阈值 `LINE_THRESHOLD`。
4. 编译上传后在黑线轨道上测试。


## 控制逻辑说明

- 左中右三个传感器先转换成在线 / 离线状态。
- 根据左右偏差计算误差。
- 通过左右轮速度差修正方向。
- 三路都丢线时停车。


## 验证标准

能沿黑线前进，弯道处能修正方向，完全丢线时能停车。
