# GPS 定位与轨迹记录系统

## 项目简介

本项目实现一个基于 ESP32 的 GPS 定位与轨迹记录系统，使用 GPS 模块获取经纬度、速度和 UTC 时间，使用 SD 卡持续记录轨迹点，并通过 OLED 显示当前位置、速度、卫星数以及日志状态。

该系统适合移动监测节点、简易车载记录器和定位课设原型，重点覆盖定位获取、速度显示、轨迹存储和时间同步四项核心需求。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| GPS 模块 | 1 | 获取经纬度和时间 |
| SD 卡模块 | 1 | 轨迹数据存储 |
| Micro SD 卡 | 1 | 保存 `track.csv` |
| 0.96 寸 I2C OLED | 1 | 显示定位状态 |
| 杜邦线 | 若干 | 原型连接 |


## 核心功能

- 获取经纬度。
- 获取速度。
- 获取 GPS 时间。
- 将轨迹点周期性写入 SD 卡。
- OLED 分页显示定位和日志状态。
- 自动创建 `track.csv` 表头。

## 运行方式

1. 使用 Arduino IDE 打开 `src/gps_positioning_and_track_logging_system/gps_positioning_and_track_logging_system.ino`。
2. 安装 `TinyGPSPlus`、`SD`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 开发板选择 `ESP32`。
4. 完成接线并插入 SD 卡后编译上传。
5. 在开阔环境等待 GPS 完成首次定位。


## 控制逻辑说明

- GPS 串口持续喂给 TinyGPSPlus 做增量解析。
- 当定位有效时，周期性把日期、时间、经纬度、速度和卫星数写入 `track.csv`。
- 第一次写文件时自动补上 CSV 表头，方便后续导出分析。
- OLED 两页轮播：位置页看实时定位，状态页看时间、SD 和 Fix 状态。
- 如果 SD 打开失败，系统会把 `sdReady` 标为异常并在显示中提示。
