# 多自由度机械臂控制系统

## 项目简介

本项目实现一个基于 Arduino Uno 的多自由度机械臂控制系统，支持底座、肩部、肘部和夹爪四个舵机通道，通过串口命令进行单关节控制，并支持动作录制与动作回放，为后续上位机、摇杆或视觉引导控制奠定基础。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 舵机 | 4 | 机械臂各关节 |
| 机械臂结构件 | 1 套 | 底座和连杆 |
| 串口终端 / 上位机 | 1 | 发送控制命令 |


## 核心功能

- 多关节舵机控制。
- 串口设置角度。
- 动作录制。
- 动作回放。
- 机械臂复位。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| BASE | D3 |
| SHOULDER | D5 |
| ELBOW | D6 |
| GRIP | D9 |


## 接线说明

四路舵机信号线分别接 D3、D5、D6、D9。舵机数量较多时请使用独立 5V 电源，并确保与 Arduino 共地。


## 运行方式

1. 打开 `src/multi_degree_of_freedom_robotic_arm_control_system/multi_degree_of_freedom_robotic_arm_control_system.ino`。
2. 接好四个舵机并确保供电稳定。
3. 编译上传后使用串口发送 `SET`、`SAVE`、`PLAY` 等命令测试。


## 控制逻辑说明

- `SET idx angle` 控制单个关节角度。
- `SAVE` 把当前四个关节角度保存为一帧动作。
- `PLAY` 按保存顺序逐帧回放。
- 角度写入前统一做机械限位约束。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 单关节控制 | 指定关节能转到目标角度 |
| 动作录制 | `SAVE` 后记录帧数增加 |
| 动作回放 | `PLAY` 后机械臂按顺序执行动作 |
| 复位功能 | `RESET` 后回到初始姿态 |
