# 简易气象站

## 项目简介

本项目实现一个基于 ESP32 的简易气象站，使用 DHT22 检测温湿度，使用 BMP280 检测气压，使用雨滴传感器检测降雨情况，使用风速脉冲传感器估算风速，并在 OLED 上分页显示结果。

该系统适合课程设计中的基础气象监测原型，也可以作为后续联网气象站的本地采集端。相比只做单一温湿度采集，这个版本把降雨和风速也纳入进来，更接近“完整气象节点”的形态。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| DHT22 | 1 | 温湿度检测 |
| BMP280 | 1 | 气压检测 |
| 雨滴传感器 | 1 | 降雨检测 |
| 风速脉冲传感器 | 1 | 风速估算 |
| 0.96 寸 I2C OLED | 1 | 信息显示 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 温度采集。
- 湿度采集。
- 气压采集。
- 雨滴数字检测和模拟强度显示。
- 风速脉冲计数与速度估算。
- OLED 分页显示总览、降雨和风速信息。

## 运行方式

1. 使用 Arduino IDE 打开 `src/simple_weather_station/simple_weather_station.ino`。
2. 安装 `DHT sensor library`、`Adafruit_BMP280`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 开发板选择 `ESP32`。
4. 完成接线后编译并上传程序。
5. 根据风速传感器实际参数调整换算系数 `WIND_FACTOR_KMH`。


## 控制逻辑说明

- 系统周期采集温湿度、气压、雨滴和风速数据。
- 雨滴模拟量做一阶低通平滑，用于观察降雨强度趋势。
- 风速使用固定时间窗内的脉冲增量换算，得到近似风速值。
- OLED 自动轮播总览页、降雨页和风速页，兼顾运行观察与标定调试。
- BMP280 初始化失败时会继续重试，减少因上电时序导致的永久故障。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 温湿度采集 | OLED 正常显示温度和湿度 |
| 气压采集 | BMP280 工作正常时显示气压值 |
| 雨滴检测 | 遇水后降雨状态和模拟值明显变化 |
| 风速统计 | 风速传感器转动时速度值发生变化 |
| 页面轮播 | 三个页面能自动切换 |
| 重试逻辑 | BMP280 初始化失败后后续能重新尝试连接 |


## 可扩展方向

- 接入 WiFi，做网页端实时气象看板。
- 增加风向检测，扩展到更完整的气象站。
- 增加日夜统计和趋势分析。
- 增加太阳能和电池，扩展为户外低功耗节点。
