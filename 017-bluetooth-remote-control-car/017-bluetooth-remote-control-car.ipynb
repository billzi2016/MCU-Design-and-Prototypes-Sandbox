# 蓝牙遥控小车

## 项目简介

本项目实现一个基于 Arduino Uno、HC-05 和 L298N 的蓝牙遥控小车，手机通过蓝牙发送简单字符命令，控制小车前进、后退、左转、右转和停止，并支持 0-9 档速度调节。

该项目适合作为嵌入式、自动化和机器人课程中的无线遥控底盘原型，也可以作为后续避障、巡线和摄像头小车的基础运动平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| HC-05 | 1 | 蓝牙串口模块 |
| L298N | 1 | 双电机驱动 |
| 直流减速电机 | 2 | 左右驱动轮 |
| 小车底盘 | 1 | 机械结构 |
| 电池组 | 1 | 电机供电 |
| 状态 LED | 1 | 运动状态指示 |


## 核心功能

- 蓝牙接收运动命令。
- 前进、后退、左转、右转、停止控制。
- 0-9 档速度调节。
- 蓝牙失联超时自动停车。
- 状态 LED 指示车辆是否处于运动状态。


## 蓝牙命令定义

- `F`：前进
- `B`：后退
- `L`：左转
- `R`：右转
- `S`：停止
- `0-9`：速度档位

| --- | --- |

| 右电机方向 | D4、D7 |
| 状态 LED | D13 |


## 接线说明

| 模块 | Arduino Uno |
| --- | --- |
| HC-05 RX | D10 |
| HC-05 TX | D11 |
| L298N ENA | D5 |
| L298N IN1/IN2 | D2/D3 |
| L298N ENB | D6 |
| L298N IN3/IN4 | D4/D7 |
| 状态 LED | D13 |

电机供电建议使用独立电池组，驱动电源地与 Arduino GND 共地。


## 运行方式

1. 使用 Arduino IDE 打开 `src/bluetooth_remote_control_car/bluetooth_remote_control_car.ino`。
2. 开发板选择 `Arduino Uno`。
3. 按照接线说明连接 HC-05、L298N 和电机。
4. 编译并上传程序。
5. 使用蓝牙串口调试 App 发送控制命令测试小车运动。


## 控制逻辑说明

- 蓝牙串口持续读取单字符命令，并映射到小车状态机。
- 方向状态和速度档位分开处理，便于用少量命令实现完整运动控制。
- 超过超时时间未收到新命令时，小车自动停车，防止蓝牙断连后继续失控运行。
- 电机输出统一通过 `applyMotorState()` 下发，减少状态分散控制。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 前进控制 | 发送 `F` 后小车前进 |
| 后退控制 | 发送 `B` 后小车后退 |
| 转向控制 | 发送 `L/R` 后小车对应转向 |
| 停止控制 | 发送 `S` 后小车停止 |
| 速度控制 | 发送 `0-9` 后速度明显变化 |
| 失联保护 | 长时间无命令输入后小车自动停车 |
