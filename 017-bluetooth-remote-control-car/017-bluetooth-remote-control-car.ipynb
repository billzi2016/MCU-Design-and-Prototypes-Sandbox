# 蓝牙遥控小车

## 项目简介

本项目实现一个基于 Arduino Uno、HC-05 和 L298N 的蓝牙遥控小车，手机通过蓝牙发送简单字符命令，控制小车前进、后退、左转、右转和停止，并支持 0-9 档速度调节。


## 核心命令

- `F`：前进
- `B`：后退
- `L`：左转
- `R`：右转
- `S`：停止
- `0-9`：速度档位


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| HC-05 | 1 | 蓝牙串口模块 |
| L298N | 1 | 双电机驱动 |
| 直流减速电机 | 2 | 左右驱动轮 |
| 小车底盘 | 1 | 机械结构 |
| 电池组 | 1 | 电机供电 |
