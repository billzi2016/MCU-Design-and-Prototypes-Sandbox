# 智能仓储分拣原型系统

## 项目简介

本项目实现一个将传送带、RFID、颜色识别、机械臂和 OLED 状态显示整合在一起的智能仓储分拣原型系统。货物进入检测位后，系统先停止传送带，读取 RFID 标签和颜色信息，再根据业务优先级和颜色类别决定投放区，最后由机械臂完成抓取与分拣。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 传送带电机与继电器 | 1 套 | 控制输送启停 |
| RFID RC522 | 1 | 读取货物标签 |
| TCS3200 类颜色传感器 | 1 | 颜色识别 |
| 舵机 | 3 | 机械臂底座、升降、夹爪 |
| OLED 显示屏 | 1 | 显示分拣状态 |
| 入料检测开关 / 红外 | 1 | 检测货物到位 |


## 核心功能

- 货物到位后自动停下传送带。
- RFID 读取货物类别或业务属性。
- 颜色识别用于辅助分类或规则分流。
- 机械臂执行抓取、旋转、放置和复位。
- OLED 实时显示当前状态、标签类型、颜色和目标分拣区。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 入料检测 | D2 |
| 传送带继电器 | D3 |
| 机械臂底座 / 升降 / 夹爪 | D5、D6、D9 |
| RFID SS / RST | D10、D8 |
| TCS3200 S0 / S1 / S2 / S3 | A0、A1、A2、A3 |
| TCS3200 OUT | A4 |
| OLED SDA / SCL | I2C 总线 |


## 接线说明

- 传送带电机由继电器控制通断，继电器控制脚接 D3。
- RC522 使用 SPI 接口，除 `SS` 和 `RST` 外其余引脚接 Arduino 默认 SPI 总线。
- TCS3200 的 `S0-S3` 接 A0-A3，`OUT` 接 A4，用于读取颜色响应脉宽。
- 三路舵机建议独立供电，并与 Arduino 共地，避免机械臂动作时 OLED 或 RFID 重启。
- 入料检测可用微动开关、对射红外或接近传感器，代码中默认低电平表示有货到位。


## 运行方式

1. 上传 `src/smart_warehouse_sorting_prototype_system/smart_warehouse_sorting_prototype_system.ino`。
2. 完成传送带、机械臂、RFID、颜色传感器和 OLED 接线。
3. 先测试传送带启停，再测试机械臂基础动作和回原点。
4. 根据实物颜色和环境光重新调整颜色判断规则。
5. 放置带 RFID 标签的货物，观察状态机是否完整走完识别、抓取和投放流程。


## 控制逻辑说明

- 系统状态分为 `WAIT_ITEM / IDENTIFY_ITEM / PICK_ITEM / PLACE_ITEM / SORT_COMPLETE`。
- 检测到货物后先停传送带，再等待一小段稳定时间避免刚到位时读卡失败。
- RFID 用于判断货物的业务属性，例如 VIP、普通件、退货件。
- 颜色识别用于辅助分类，最终由 `decideTargetZone()` 综合得到目标分拣区。
- 分拣完成后系统清空当前货物信息、重新启动传送带并等待下一件货物。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 到位停带 | 货物进入检测位后传送带停止 |
| RFID 识别 | OLED 和串口能显示标签类别 |
| 颜色识别 | 红绿蓝或混合色能被区分 |
| 分拣执行 | 机械臂能将货物放到目标分拣区 |


## 可扩展方向

- 增加数据库记录和仓库件号管理。
- 增加视觉识别，替代简单颜色分类。
- 增加上位机任务调度和分拣统计。
- 增加重量检测或条码识别形成多特征分类。
