# 多功能数码管电子钟

## 项目简介

这是一个基于 Arduino Uno 的多功能数码管电子钟原型系统，用于实现时间显示、时间校准、闹钟提醒和整点蜂鸣提示。

系统使用 DS3231 实时时钟模块保存准确时间，使用 TM1637 四位数码管显示小时和分钟，使用按键完成模式切换和参数调整。Arduino Uno 的引脚数量和性能足够，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| DS3231 RTC 模块 | 1 | 实时时钟，断电后可保持时间 |
| TM1637 四位数码管模块 | 1 | 显示时间和设置参数 |
| 轻触按键 | 4 | 模式、加、减、确认 |
| 有源蜂鸣器 | 1 | 闹钟和整点提示 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 显示当前小时和分钟。
- 数码管中间冒号按秒闪烁。
- 支持时间设置。
- 支持闹钟小时和分钟设置。
- 支持闹钟开关。
- 到达闹钟时间后蜂鸣器提示。
- 支持整点短蜂鸣提示。
- 按键使用软件消抖，避免一次按下触发多次。

| DIO | D3 |

### 按键

程序使用 Arduino Uno 内部上拉电阻，因此每个按键一端接对应数字引脚，另一端接 GND。按键未按下时，引脚读数为 HIGH；按键按下时，引脚被拉低为 LOW。

| 按键功能 | 一端连接 | 另一端连接 |
| --- | --- | --- |
| 模式切换 | D4 | GND |
| 参数增加 | D5 | GND |
| 参数减少 | D6 | GND |
| 确认保存 | D7 | GND |

### 蜂鸣器

| 蜂鸣器 | Arduino Uno |
| --- | --- |
| 正极 | D8 |
| 负极 | GND |

当前代码按有源蜂鸣器处理。如果使用无源蜂鸣器，可以后续改为 `tone()` 输出。


## 操作说明

| 操作 | 功能 |
| --- | --- |
| 短按模式键 | 在正常显示、设置小时、设置分钟、设置闹钟小时、设置闹钟分钟、闹钟开关之间切换 |
| 短按加键 | 当前设置项加 1 |
| 短按减键 | 当前设置项减 1 |
| 短按确认键 | 保存当前设置并返回正常显示 |


## 运行方式

1. 使用 Arduino IDE 打开 `src/digital_clock/digital_clock.ino`。
2. 安装 `RTClib` 库。
3. 安装 `TM1637Display` 库。
4. 开发板选择 `Arduino Uno`。
5. 按照接线说明完成连接。
6. 编译并上传程序。


## 状态变量

- `mode`：当前显示或设置模式。
- `settingHour`：设置中的小时。
- `settingMinute`：设置中的分钟。
- `alarmHour`：闹钟小时。
- `alarmMinute`：闹钟分钟。
- `alarmEnabled`：闹钟是否启用。
- `lastAlarmMinute`：记录最近一次响铃分钟，避免同一分钟内重复触发。
