# 智能台灯控制系统

## 项目简介

本项目实现一个基于 Arduino Uno 的智能台灯控制系统，使用光敏电阻检测环境亮度，使用 PWM 调节 LED 灯亮度，并通过按键完成自动模式、手动模式和亮度调整切换。

系统使用 I2C OLED 显示当前工作模式、环境亮度值和目标亮度值。Arduino Uno 可以满足光照采集、PWM 输出、按键输入和显示控制需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 光敏电阻 | 1 | 检测环境光照强度 |
| 10k 电阻 | 1 | 与光敏电阻组成分压电路 |
| LED 灯或 LED 模块 | 1 | 台灯光源 |
| NPN 三极管或 MOSFET | 1 | 放大 PWM 驱动能力 |
| 220Ω 电阻 | 1 | LED 限流，按灯珠规格调整 |
| 0.96 寸 I2C OLED | 1 | 显示模式和亮度信息 |
| 轻触按键 | 3 | 模式切换、亮度增加、亮度减少 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 光敏电阻实时采集环境亮度。
- 自动模式下根据环境亮度自动调节灯光亮度。
- 手动模式下使用按键增加或减少亮度。
- 模式按键在自动模式和手动模式之间切换。
- OLED 实时显示模式、环境亮度、输出亮度。
- 对模拟采样值做平滑处理，避免亮度跳变过快。
- 按键使用软件消抖，避免一次按下触发多次。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 光敏电阻分压输出 | A0 |
| 模式切换按键 | D2 |
| 亮度增加按键 | D3 |
| 亮度减少按键 | D4 |
| LED PWM 输出 | D5 |
| OLED SDA | A4 |
| OLED SCL | A5 |


## 接线说明

### 光敏电阻接线

光敏电阻和 10k 电阻组成分压电路，分压中点接到 A0。

| 元件连接 | 说明 |
| --- | --- |
| 光敏电阻一端 | 5V |
| 光敏电阻另一端 | A0 和 10k 电阻一端 |
| 10k 电阻另一端 | GND |

### 按键接线

程序使用 Arduino Uno 内部上拉电阻，因此每个按键一端接对应数字引脚，另一端接 GND。按键未按下时读数为 HIGH，按下时读数为 LOW。

| 按键功能 | 一端连接 | 另一端连接 |
| --- | --- | --- |
| 模式切换 | D2 | GND |
| 亮度增加 | D3 | GND |
| 亮度减少 | D4 | GND |

### LED 灯驱动接线

如果使用单个小功率 LED，可直接通过限流电阻接 PWM 引脚。如果使用功率更高的 LED 模块，建议通过三极管或 MOSFET 驱动。

推荐使用三极管驱动方式：

| 元件 | 连接 |
| --- | --- |
| D5 | 通过 1k 电阻接三极管基极 |
| 三极管发射极 | GND |
| 三极管集电极 | LED 负极 |
| LED 正极 | 通过限流电阻接 5V |

### OLED 接线

| OLED | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| SDA | A4 |
| SCL | A5 |


## 运行方式

1. 使用 Arduino IDE 打开 `src/smart_desk_lamp_controller/smart_desk_lamp_controller.ino`。
2. 安装 `Adafruit_GFX` 库。
3. 安装 `Adafruit_SSD1306` 库。
4. 开发板选择 `Arduino Uno`。
5. 按照接线说明完成连接。
6. 编译并上传程序。


## 状态变量

- `autoMode`：当前是否为自动模式。
- `lightRaw`：光敏电阻原始采样值。
- `lightFiltered`：平滑后的环境亮度值。
- `targetBrightness`：最终输出亮度。
- `manualBrightness`：手动模式下的亮度设定值。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 自动调光 | 环境变暗时 LED 亮度增大，环境变亮时亮度减小 |
| 模式切换 | 按模式键后可在自动和手动模式之间切换 |
| 手动加亮 | 手动模式下按增加键可提升亮度 |
| 手动减亮 | 手动模式下按减少键可降低亮度 |
| OLED 显示 | 屏幕正确显示模式、环境亮度和输出亮度 |
| 按键消抖 | 单次按下只触发一次操作 |
| 平滑处理 | 环境光变化时 LED 亮度不会剧烈跳变 |


## 可扩展方向

- 增加人体红外感应，实现无人自动熄灯。
- 增加 RTC，实现定时开关灯。
- 增加触摸按键或旋钮调节。
- 接入 WiFi，实现远程控制。
- 改用恒流驱动方案控制更高功率灯源。
