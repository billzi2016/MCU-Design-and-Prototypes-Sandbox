# GSM/SMS 短信报警系统

## 项目简介

本项目实现一个基于 Arduino Uno 与 SIM800L 的短信报警系统，使用开关量传感器检测异常状态，通过短信向指定手机号发送报警信息，并接收短信命令完成布防、撤防、警报器开关和状态查询。

该实现把核心重点放在异常检测、短信发送、短信指令控制和状态反馈，适合作为后续烟雾、水位、门磁等报警系统的通信扩展模块。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| SIM800L | 1 | GSM/SMS 通信模块 |
| 门磁或开关量传感器 | 1 | 异常检测输入 |
| 有源蜂鸣器 | 1 | 本地报警提示 |
| 外部稳压电源 | 1 | 为 SIM800L 供电 |
| 杜邦线 | 若干 | 原型连接 |


## 支持的短信指令

- `STATUS`：查询当前状态。
- `ARM`：布防。
- `DISARM`：撤防。
- `SIREN ON`：开启本地蜂鸣器报警。
- `SIREN OFF`：关闭本地蜂鸣器报警。


## 运行方式

1. 使用 Arduino IDE 打开 `src/gsm_sms_alert_system/gsm_sms_alert_system.ino`。
2. 确认 `AUTHORIZED_PHONE` 修改为实际手机号。
3. 给 SIM800L 提供稳定电源，并插入可用 SIM 卡。
4. 编译并上传程序。
