# GSM/SMS 短信报警系统

## 项目简介

本项目实现一个基于 Arduino Uno 与 SIM800L 的短信报警系统，使用门磁或开关量传感器检测异常状态，通过短信向指定手机号发送报警信息，并接收短信命令完成布防、撤防、警号开关和状态查询。

该实现把重点放在异常检测、短信发送、短信指令控制和状态反馈上，适合作为后续烟雾、水位、门禁等报警项目的通信扩展模块。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| SIM800L | 1 | GSM/SMS 通信模块 |
| 门磁或开关量传感器 | 1 | 异常检测输入 |
| 有源蜂鸣器 | 1 | 本地报警提示 |
| 外部稳压电源 | 1 | 为 SIM800L 供电 |
| 杜邦线 | 若干 | 原型连接 |


## 核心功能

- 传感器异常检测。
- 报警短信发送。
- 恢复短信发送。
- 本地蜂鸣器报警。
- 短信指令远程布防、撤防和状态查询。


## 支持的短信指令

- `STATUS`：查询当前状态。
- `ARM`：布防。
- `DISARM`：撤防。
- `SIREN ON`：开启本地蜂鸣器报警。
- `SIREN OFF`：关闭本地蜂鸣器报警。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 传感器输入 | D2 |
| 蜂鸣器 | D8 |
| SIM800L RX | D10 |
| SIM800L TX | D11 |


## 接线说明

### 传感器与蜂鸣器接线

| 模块 | Arduino Uno |
| --- | --- |
| 门磁 / 开关量信号 | D2 |
| 门磁 / 开关量另一端 | GND |
| 蜂鸣器正极 | D8 |
| 蜂鸣器负极 | GND |

程序使用 `INPUT_PULLUP`，因此触发时输入通常为 `LOW`。

### SIM800L 接线

| SIM800L | Arduino Uno |
| --- | --- |
| TXD | D10 |
| RXD | D11 |
| GND | GND |
| VCC | 外部稳压电源 |

SIM800L 电流峰值较大，不建议直接由 Arduino 5V 引脚供电。


## 运行方式

1. 使用 Arduino IDE 打开 `src/gsm_sms_alert_system/gsm_sms_alert_system.ino`。
2. 确认 `AUTHORIZED_PHONE` 修改为实际手机号。
3. 给 SIM800L 提供稳定电源，并插入可用 SIM 卡。
4. 完成接线后编译并上传程序。
5. 上电后用手机发送 `STATUS` 测试收发链路。


## 控制逻辑说明

- 本地输入先做消抖，再判断是否进入报警状态。
- 布防状态下触发异常时，系统发送报警短信并启用本地蜂鸣器。
- 异常恢复后补发恢复短信，便于远程确认现场已恢复正常。
- 短信指令通过 SIM800L 直推文本行方式解析，便于用最简单的 AT 流程实现控制。
- 蜂鸣器是短信报警的补充输出，可远程单独启停。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 触发检测 | 传感器触发后系统进入报警状态 |
| 报警短信 | 指定手机号收到报警短信 |
| 恢复短信 | 异常恢复后收到恢复短信 |
| 本地蜂鸣器 | 报警时蜂鸣器按节奏鸣叫 |
| 远程指令 | `STATUS`、`ARM`、`DISARM` 等指令可生效 |
| 撤防状态 | 撤防后传感器触发不再进入报警流程 |


## 可扩展方向

- 增加多路报警输入，扩展为多区域告警系统。
- 增加密码验证，提升短信指令安全性。
- 与烟雾、水位、门禁等项目联动。
- 增加事件记录功能，保存最近一次报警时间和原因。
- 扩展为 GSM 来电告警或语音播报。
