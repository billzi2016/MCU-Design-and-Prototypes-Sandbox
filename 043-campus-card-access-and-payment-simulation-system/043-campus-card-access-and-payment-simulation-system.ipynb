# 校园一卡通门禁与消费模拟系统

## 项目简介

本项目实现一个基于 RFID、LCD、蜂鸣器和 EEPROM 的校园一卡通门禁与消费模拟系统。系统通过识别不同卡号，完成门禁放行、余额显示、消费扣费和管理员充值等操作，适合作为门禁和消费子系统合并的校园场景原型。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| RC522 RFID | 1 | 卡片识别 |
| LCD1602 I2C | 1 | 显示状态和余额 |
| 蜂鸣器 | 1 | 操作反馈 |
| 继电器 / 门锁模拟输出 | 1 | 门禁放行 |
| 按键 | 2 | 消费模式和充值模式 |
| EEPROM | 内置 | 保存卡片余额与角色 |


## 核心功能

- 刷卡识别用户身份。
- 默认模式下执行门禁放行。
- 消费模式下刷卡扣费并显示余额。
- 管理员卡可在充值模式下完成加值。
- EEPROM 保存卡片余额和角色，支持断电后保留。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| RC522 SS / RST | D10、D9 |
| 蜂鸣器 | D8 |
| 门禁继电器 | D7 |
| 消费模式按键 | D6 |
| 充值模式按键 | D5 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- RC522 使用 SPI 接口，除 `SS` 和 `RST` 外，其余按默认 SPI 口连接。
- 继电器接 D7，用于模拟门锁执行回路。
- 消费和充值模式按键使用内部上拉，按下有效。
- LCD 通过 I2C 与主控连接，便于实时显示卡号片段和余额。


## 运行方式

1. 上传 `src/campus_card_access_and_payment_simulation_system/campus_card_access_and_payment_simulation_system.ino`。
2. 上电后会初始化两张默认卡：学生卡和管理员卡。
3. 默认模式刷卡测试门禁放行。
4. 按住消费模式按键再刷卡，测试扣费和余额变化。
5. 按住充值模式按键再刷管理员卡，测试充值功能。


## 控制逻辑说明

- 程序先在 EEPROM 中查找匹配 UID 的卡记录。
- 普通模式下只做门禁放行，不扣减余额。
- 消费模式下先检查余额是否充足，再执行扣费并写回 EEPROM。
- 充值模式仅允许管理员卡操作，避免任意卡片加值。
- LCD 与蜂鸣器共同反馈当前结果，包括未知卡、余额不足、扣费成功和充值成功等情况。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 门禁识别 | 已登记卡可触发继电器开门 |
| 消费扣费 | 消费模式下余额会减少 |
| 余额不足 | 余额不足时拒绝扣费 |
| 管理员充值 | 管理员卡可增加余额并写回 EEPROM |


## 可扩展方向

- 增加多用户卡片管理界面。
- 增加上位机日志与消费记录导出。
- 增加权限分组和不同消费场景。
- 接入真实键盘实现商品或金额选择。
