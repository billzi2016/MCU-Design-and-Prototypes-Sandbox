# 智能停车场车位检测系统

## 项目简介

本项目实现一个基于多路超声波、LCD、LED 和舵机道闸的智能停车场车位检测系统。系统实时监测 3 个车位的占用状态，统计剩余车位数量，并依据入口 / 出口请求控制道闸开闭，适合作为停车场原型或课程设计演示平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| HC-SR04 | 3 | 检测三个车位占用 |
| 舵机 | 1 | 道闸控制 |
| LCD1602 I2C | 1 | 显示剩余车位与状态 |
| 绿灯 / 红灯 | 各 1 | 入口状态指示 |
| 入口 / 出口按键 | 各 1 | 模拟车辆进出请求 |


## 核心功能

- 多车位占用检测。
- 剩余车位实时统计。
- 停车场入口满位 / 可进状态指示。
- 道闸自动开闭控制。
- LCD 显示车位占用图和道闸状态。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 道闸舵机 | D3 |
| 入口按钮 / 出口按钮 | D4、D5 |
| 绿灯 / 红灯 | D6、D7 |
| 车位 1 Trig / Echo | D8、D9 |
| 车位 2 Trig / Echo | D10、D11 |
| 车位 3 Trig / Echo | D12、D13 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- 每个车位上方或前方安装一个超声波模块，确保对准停车位置。
- 舵机连接道闸杆机构，角度 `10` 度为关闭，`95` 度为抬起示意。
- 入口和出口按钮模拟车辆到达感应与离场请求。
- 绿灯用于显示可进入，红灯用于显示满位或禁止进入。


## 运行方式

1. 上传 `src/smart_parking_lot_space_detection_system/smart_parking_lot_space_detection_system.ino`。
2. 分别测试三个超声波模块的测距结果，并调整安装高度。
3. 根据实际车位大小微调 `OCCUPIED_DISTANCE_CM`。
4. 按入口按钮测试有空位时的开闸逻辑。
5. 按出口按钮测试离场开闸逻辑和车位恢复显示。


## 控制逻辑说明

- 程序周期采样三个车位的距离，并据此判断是否占用。
- 空位数量实时刷新到 LCD，同时影响红绿灯显示。
- 入口按钮只有在仍有空位时才允许开闸。
- 出口按钮不受空位数量限制，用于模拟车辆正常离场。
- 道闸打开后保持一段时间自动关闭，减少按钮长按或忘记关闸的问题。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 车位检测 | 三个车位占用状态显示正确 |
| 空位统计 | 剩余车位数量会随占用变化 |
| 入口控制 | 有空位时入口按钮可开闸 |
| 满位提示 | 满位时红灯亮且入口不放行 |
