# 迷你自动售货机原型

## 项目简介

本项目实现一个基于 Arduino、按钮、LCD、舵机和 RFID 支付模拟的迷你自动售货机原型。用户通过按钮选择商品，刷卡完成余额支付后，系统驱动舵机出货，并同步更新库存和余额信息，适合作为支付控制与库存管理结合的综合小项目。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| RC522 RFID | 1 | 模拟支付卡识别 |
| LCD1602 I2C | 1 | 显示商品、余额和库存 |
| 舵机 | 2 | 模拟两个商品通道出货 |
| 按键 | 3 | 商品选择与购买确认 |
| 蜂鸣器 | 1 | 操作反馈 |


## 核心功能

- 按钮选择两类商品。
- 刷卡模拟支付余额验证。
- 支付成功后驱动对应舵机出货。
- 库存数量和卡余额实时更新。
- 对缺货、余额不足和未刷卡等场景给出明确提示。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 商品 A 选择 | D2 |
| 商品 B 选择 | D3 |
| 购买确认 | D4 |
| 商品 A 舵机 | D5 |
| 商品 B 舵机 | D6 |
| 蜂鸣器 | D8 |
| RC522 SS / RST | D10、D9 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- RC522 按 SPI 方式连接，`SS` 接 D10，`RST` 接 D9。
- 两个舵机分别对应两个商品出货通道。
- 三个按键使用内部上拉输入，按下有效。
- LCD 通过 I2C 接口显示当前商品、库存和余额。
- 若后续改为硬币器或扫码模块，只需替换支付入口逻辑。


## 运行方式

1. 上传 `src/mini_vending_machine_prototype/mini_vending_machine_prototype.ino`。
2. 连接 RC522、两路舵机、按钮、蜂鸣器和 LCD。
3. 使用按钮选择商品 A 或 B。
4. 按购买键后刷卡，验证支付、出货和库存变化。
5. 分别测试缺货、余额不足和未刷卡场景。


## 控制逻辑说明

- 系统先通过选择按钮确定当前商品通道。
- 购买时优先检查库存，再检查是否刷卡与余额是否足够。
- 支付成功后扣减余额和库存，并驱动对应舵机旋转一次模拟出货。
- 两个商品通道独立维护价格和库存，便于扩展更多品类。
- LCD 与蜂鸣器共同承担交易反馈，减少仅靠串口观察的依赖。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 商品选择 | 可在两种商品之间切换 |
| 支付扣费 | 刷卡后余额会按商品价格减少 |
| 舵机出货 | 支付成功后对应舵机动作一次 |
| 库存管理 | 出货后库存减少，缺货时拒绝购买 |


## 可扩展方向

- 增加更多商品通道和库存管理页面。
- 增加管理员补货模式。
- 接入扫码支付或上位机交易记录。
- 增加异常卡片、退款和维护模式。
