# GPS 测距与电子围栏系统

## 项目简介

本项目实现一个基于 ESP32 的 GPS 测距与电子围栏系统，使用 GPS 模块获取当前位置，计算与目标点之间的距离，并根据围栏半径判断是否越界。系统使用 OLED 显示距离、卫星数和围栏状态，使用蜂鸣器在越界或 GPS 信号丢失时发出提示。

该项目适合做地理围栏、巡检路线管理、移动设备越界告警等课设原型，也可以作为后续远程定位项目的本地告警端。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| GPS 模块 | 1 | 获取经纬度 |
| 0.96 寸 I2C OLED | 1 | 显示距离和状态 |
| 有源蜂鸣器 | 1 | 越界报警 |
| 杜邦线 | 若干 | 原型连接 |


## 核心功能

- 获取当前位置经纬度。
- 计算与目标点的距离。
- 判断是否超出电子围栏半径。
- 越界或 GPS 丢失时蜂鸣器报警。
- OLED 分页显示位置和围栏结果。

## 运行方式

1. 使用 Arduino IDE 打开 `src/gps_ranging_and_geofence_system/gps_ranging_and_geofence_system.ino`。
2. 安装 `TinyGPSPlus`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 开发板选择 `ESP32`。
4. 根据需要修改目标点经纬度和围栏半径。
5. 完成接线后编译并上传程序。


## 控制逻辑说明

- GPS 串口字节流持续喂入 TinyGPSPlus 做解析。
- 当位置有效时，使用 Haversine 公式计算当前位置到目标点的球面距离。
- 若距离超出围栏半径，系统进入越界状态；若长时间无有效定位，则进入 GPS 丢失告警状态。
- 蜂鸣器在越界或 GPS 丢失时工作，正常时保持静音。
- OLED 两页轮播显示位置页和围栏页。
