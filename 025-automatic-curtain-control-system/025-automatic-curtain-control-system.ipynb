# 自动窗帘控制系统

## 项目简介

本项目实现一个基于 Arduino Uno 的自动窗帘控制系统，使用步进电机驱动窗帘开合，结合 RTC 时间、光敏电阻和限位开关进行自动控制，同时支持按键手动开合。系统适合作为智能家居自动执行机构的基础原型。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 步进电机 | 1 | 窗帘驱动 |
| RTC 模块 | 1 | 定时依据 |
| 光敏电阻 | 1 | 光照检测 |
| 限位开关 | 2 | 开到位 / 关到位保护 |
| 按键 | 1 | 手动切换 |


## 核心功能

- 定时开合窗帘。
- 光照触发开合。
- 手动按键控制。
- 限位保护。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 步进电机四相 | D8、D10、D9、D11 |
| 光敏电阻 | A0 |
| 开到位限位 | D2 |
| 关到位限位 | D3 |
| 手动按键 | D4 |


## 接线说明

步进电机四相按代码定义顺序接入；光敏电阻接 A0；两个限位开关分别接 D2、D3；按键使用内部上拉输入接 D4。RTC 使用默认 I2C 接口连接。


## 运行方式

1. 打开 `src/automatic_curtain_control_system/automatic_curtain_control_system.ino`。
2. 安装 `RTClib` 库。
3. 完成步进电机、RTC、光敏电阻、限位和按键接线。
4. 编译上传后校准开合时间和阈值。


## 控制逻辑说明

- 系统定期读取 RTC 时间和环境光照。
- 白天且光照充足时尝试开窗帘。
- 夜晚或光照不足时尝试关窗帘。
- 手动按钮可直接切换开关状态，限位开关负责行程保护。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 定时开合 | 指定时间段内窗帘按逻辑开合 |
| 光照触发 | 遮挡或增强光照后动作逻辑变化 |
| 手动控制 | 按键可切换窗帘状态 |
| 限位保护 | 到位后电机停止，不继续强拉 |
