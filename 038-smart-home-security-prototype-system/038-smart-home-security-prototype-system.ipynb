# 智能家居安防原型系统

## 项目简介

本项目实现一个基于 ESP32 的智能家居安防原型系统，集成温湿度监测、烟雾检测、门磁检测、人体红外入侵检测、本地蜂鸣报警和继电器联动。系统支持布防与撤防切换，并对环境异常和入侵事件使用不同的报警逻辑，适合作为综合课设型原型。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 | 1 | 主控平台 |
| DHT22 | 1 | 温湿度采集 |
| MQ-2 | 1 | 烟雾检测 |
| 门磁开关 | 1 | 门状态检测 |
| PIR 人体红外 | 1 | 入侵感知 |
| 蜂鸣器、继电器 | 各 1 | 告警与联动 |
| I2C LCD1602 | 1 | 本地状态显示 |


## 核心功能

- 周期采集温湿度与烟雾数据。
- 布防状态下监测门磁与人体红外，判断是否存在入侵行为。
- 支持环境告警和入侵告警两类不同报警源。
- LCD 轮播显示环境数据、门磁状态和系统模式。
- 通过按键实现布防 / 撤防和告警确认。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| DHT22 | GPIO4 |
| MQ-2 模拟输出 | GPIO34 |
| 门磁 | GPIO18 |
| PIR | GPIO19 |
| 蜂鸣器 | GPIO25 |
| 继电器 | GPIO26 |
| 布防按键 | GPIO27 |
| 确认按键 | GPIO14 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- DHT22 数据脚接 GPIO4，MQ-2 模拟输出接 GPIO34。
- 门磁建议使用 `INPUT_PULLUP` 接法，一端接 GND，一端接 GPIO18。
- PIR 输出接 GPIO19，蜂鸣器和继电器分别接 GPIO25 和 GPIO26。
- 布防和确认按键接 GPIO27、GPIO14，并使用内部上拉。
- LCD1602 通过 I2C 接入 ESP32，需确认地址与驱动库配置一致。


## 运行方式

1. 上传 `src/smart_home_security_prototype_system/smart_home_security_prototype_system.ino` 到 ESP32。
2. 完成 DHT22、MQ-2、门磁、PIR、蜂鸣器、继电器和 LCD 接线。
3. 根据环境和传感器特性调整烟雾和温度阈值。
4. 上电后先在撤防模式下观察环境数据，再切换到布防模式测试入侵报警。
5. 使用确认按键测试告警解除流程和保持逻辑。


## 控制逻辑说明

- 系统维护 `DISARMED / ARMED / ALARM` 三种模式。
- 撤防模式下只做环境监测，不对门磁和 PIR 触发入侵报警。
- 布防模式下门磁或人体红外触发后进入入侵报警，并保持一段锁存时间。
- 烟雾和高温异常会触发环境告警，与入侵告警共同驱动蜂鸣器和继电器。
- 蜂鸣器根据环境告警和入侵告警采用不同节奏，便于快速区分报警来源。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 环境监测 | LCD 能稳定显示温湿度与烟雾值 |
| 布防切换 | 按键可在撤防与布防之间切换 |
| 入侵报警 | 布防时门磁或 PIR 触发会进入告警状态 |
| 环境报警 | 烟雾或温度超限时蜂鸣器和继电器动作 |
