# 交通灯控制系统

## 项目简介

本项目实现一个基于 Arduino Uno 的交通灯控制系统，包含机动车红黄绿灯、人行道红绿灯、行人请求按钮和倒计时显示。系统按照固定相位运行，并在检测到行人请求后进入人行通行阶段。

Arduino Uno 可以满足 LED 控制、按键输入、数码管倒计时和状态机调度需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 红色 LED | 2 | 机动车红灯、人行道红灯 |
| 黄色 LED | 1 | 机动车黄灯 |
| 绿色 LED | 2 | 机动车绿灯、人行道绿灯 |
| TM1637 四位数码管模块 | 1 | 显示当前阶段剩余秒数 |
| 轻触按键 | 1 | 行人过街请求 |
| 有源蜂鸣器 | 1 | 行人通行提示 |
| 220Ω 电阻 | 5 | LED 限流 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 机动车红、黄、绿灯按状态机自动切换。
- 人行道红、绿灯与机动车相位互锁。
- TM1637 数码管显示当前阶段剩余秒数。
- 行人按键可以发起过街请求。
- 行人请求不会立即打断机动车绿灯，而是在安全阶段处理。
- 人行绿灯阶段蜂鸣器间歇提示。
- 所有计时使用 `millis()`，避免阻塞按键检测。


## 状态设计

| 状态 | 机动车灯 | 人行道灯 | 持续时间 |
| --- | --- | --- | --- |
| 机动车通行 | 绿灯 | 红灯 | 15 秒 |
| 机动车警示 | 黄灯 | 红灯 | 3 秒 |
| 行人通行 | 红灯 | 绿灯 | 10 秒 |
| 全红缓冲 | 红灯 | 红灯 | 2 秒 |

系统默认循环执行机动车通行、机动车警示、全红缓冲。检测到行人请求后，在机动车警示结束后进入行人通行阶段，之后再进入全红缓冲并恢复机动车通行。

## 接线说明

### LED 接线

每个 LED 的长脚接对应数字引脚，短脚通过 220Ω 电阻接 GND。

| LED | 长脚连接 | 短脚连接 |
| --- | --- | --- |
| 机动车红灯 | D2 | 220Ω 电阻后接 GND |
| 机动车黄灯 | D3 | 220Ω 电阻后接 GND |
| 机动车绿灯 | D4 | 220Ω 电阻后接 GND |
| 人行道红灯 | D5 | 220Ω 电阻后接 GND |
| 人行道绿灯 | D6 | 220Ω 电阻后接 GND |

### 按键接线

程序使用 Arduino Uno 内部上拉电阻，因此按键一端接 D7，另一端接 GND。按键未按下时读数为 HIGH，按下时读数为 LOW。

### 蜂鸣器接线

| 蜂鸣器 | Arduino Uno |
| --- | --- |
| 正极 | D8 |
| 负极 | GND |

### TM1637 数码管接线

| TM1637 | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| CLK | D9 |
| DIO | D10 |


## 运行方式

1. 使用 Arduino IDE 打开 `src/traffic_light_controller/traffic_light_controller.ino`。
2. 安装 `TM1637Display` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。

- 增加车流量检测，实现自适应绿灯时长。
