# 简易无人机姿态采集与显示系统

## 项目简介

本项目实现一个基于 MPU6050 和 OLED 的简易无人机姿态采集与显示系统。主控板周期读取姿态角数据，对读数做基础平滑滤波，并在 OLED 上实时显示 Roll、Pitch、Yaw，同时通过串口输出姿态数据，为后续无线传输、遥控调试和飞控联调提供基础传感链路。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino / STM32 | 1 | 主控板 |
| MPU6050 | 1 | IMU 姿态传感器 |
| OLED | 1 | 实时姿态显示 |
| 无线模块或蓝牙串口 | 选配 | 后续无线传输扩展 |


## 核心功能

- 周期更新 MPU6050 姿态角。
- 启动时执行零偏校准。
- 对姿态角进行基础低通平滑。
- OLED 显示实时姿态角与原始对比信息。
- 串口输出统一格式的姿态遥测数据。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| MPU6050 SDA / SCL | I2C 总线 |
| OLED SDA / SCL | I2C 总线 |
| 无线串口扩展 | 复用串口或外接 UART |


## 接线说明

- MPU6050 与 OLED 均通过 I2C 接口连接到主控板。
- 启动校准时应保持平台静止，否则零偏会被错误学习进去。
- 若后续扩展蓝牙或无线模块，可直接转发串口姿态数据。
- 供电应稳定，避免电源噪声导致 IMU 数据漂移加剧。


## 运行方式

1. 上传 `src/simple_drone_attitude_acquisition_and_display_system/simple_drone_attitude_acquisition_and_display_system.ino`。
2. 上电后保持平台静止，等待 MPU6050 校准完成。
3. 观察 OLED 是否能显示 Roll、Pitch、Yaw 三个姿态角。
4. 通过串口查看 `ATT,...` 格式数据是否连续稳定输出。
5. 轻微倾斜平台，验证滤波后姿态是否平滑变化。


## 控制逻辑说明

- 主循环以固定采样周期更新姿态，减少显示和输出节奏抖动。
- 程序同时保留原始姿态角和滤波后姿态角，便于对比观察平滑效果。
- 当前滤波采用简单一阶低通思想，重点是让原型的显示和传输更稳定。
- 串口输出采用 `ATT,roll,pitch,yaw` 统一格式，后续无线传输可直接复用。
- 该版本重点在姿态采集链路，不涉及电机混控和闭环飞控。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 姿态采集 | 倾斜平台时姿态角连续变化 |
| 校准效果 | 静止时角度漂移相对较小 |
| 显示输出 | OLED 能稳定刷新三轴姿态 |
| 遥测串口 | 串口连续输出 `ATT` 格式数据 |


## 可扩展方向

- 增加蓝牙 / LoRa / WiFi 姿态传输。
- 引入更完整的互补滤波或卡尔曼滤波。
- 增加遥控输入与飞行姿态控制闭环。
- 扩展为四旋翼飞控实验平台的传感器子模块。
