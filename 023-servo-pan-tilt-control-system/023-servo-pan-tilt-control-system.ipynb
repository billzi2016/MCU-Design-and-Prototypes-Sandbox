# 舵机云台控制系统

## 项目简介

本项目实现一个基于 Arduino Uno 的双轴舵机云台控制系统，使用双轴摇杆控制水平和俯仰两个舵机，OLED 显示当前角度，并通过限位约束保证舵机工作在安全范围内。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| SG90 舵机 | 2 | 水平 / 俯仰执行 |
| 双轴摇杆 | 1 | 控制输入 |
| 0.96 寸 I2C OLED | 1 | 角度显示 |


## 核心功能

- 双轴摇杆输入。
- 双舵机角度控制。
- 云台限位保护。
- OLED 角度显示。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 摇杆 X | A0 |
| 摇杆 Y | A1 |
| 水平舵机 | D9 |
| 俯仰舵机 | D10 |
| OLED SDA | A4 |
| OLED SCL | A5 |


## 接线说明

摇杆两个模拟轴分别接 A0、A1；两个舵机信号线接 D9、D10；OLED 使用 I2C 接口接 A4、A5。舵机建议独立供电并共地。


## 运行方式

1. 打开 `src/servo_pan_tilt_control_system/servo_pan_tilt_control_system.ino`。
2. 安装 `Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 完成摇杆、舵机和 OLED 接线。
4. 编译上传后推动摇杆观察云台动作。


## 控制逻辑说明

- 摇杆中心设置死区，过滤中位抖动。
- 水平和俯仰角度都受到软件限位约束。
- OLED 用于实时反馈当前角度，便于机械调试。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 水平控制 | 摇杆左右推动时水平舵机转动 |
| 俯仰控制 | 摇杆上下推动时俯仰舵机转动 |
| 限位保护 | 舵机不会超出设定角度范围 |
| OLED 显示 | 屏幕正确显示角度 |
