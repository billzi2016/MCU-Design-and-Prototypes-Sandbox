# 小型机械臂分拣系统

## 项目简介

本项目实现一个基于 Arduino、颜色传感器和四舵机机械臂的小型分拣系统。系统先读取颜色传感器的 RGB 响应，判断物体属于红、绿、蓝中的哪一类，再控制机械臂执行抓取、旋转放置和回原点动作，实现基础颜色分拣流程。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino 主控板 | 1 | 执行识别与动作控制 |
| TCS3200 类颜色传感器 | 1 | 读取 RGB 响应 |
| 舵机 | 4 | 控制底座、主臂、前臂、夹爪 |
| 机械臂结构 | 1 套 | 完成抓取和放置动作 |
| 分拣托盘 | 3 区 | 对应不同颜色落点 |


## 核心功能

- 依次读取颜色传感器三个颜色通道。
- 根据最强响应判断红、绿、蓝目标。
- 机械臂执行抓取动作序列。
- 不同颜色对应不同放置角度。
- 每轮结束后自动回到统一初始姿态。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 颜色传感器 S0 / S1 | D2、D3 |
| 颜色传感器 S2 / S3 | D4、D5 |
| 颜色传感器 OUT | D6 |
| 底座舵机 | D8 |
| 主臂舵机 | D9 |
| 前臂舵机 | D10 |
| 夹爪舵机 | D11 |


## 接线说明

- 颜色传感器的 `S0-S3` 分别接 D2-D5，`OUT` 接 D6，电源与主控共地。
- 四路舵机建议使用独立 5V 电源供电，信号线接对应数字口，地线与主控共地。
- 机械臂初始姿态需要保证夹爪、主臂、前臂之间不会互相碰撞。
- 分拣托盘的位置应与 `moveToDropPose()` 中的底座角度对应，必要时按实际结构微调。


## 运行方式

1. 打开 `src/small_robotic_arm_sorting_system/small_robotic_arm_sorting_system.ino`。
2. 完成颜色传感器与四路舵机接线。
3. 将待识别物体放在传感器有效检测区域。
4. 上传程序后观察串口输出的 RGB 响应值。
5. 根据实际光照和物体材质调整识别阈值与放置角度。


## 控制逻辑说明

- `detectColor()` 依次切换颜色滤波配置，分别读取红、绿、蓝三个通道的响应时间。
- 脉宽越小说明该颜色响应越强，因此选择最小值对应的颜色作为分类结果。
- 若三个颜色值都较大，则判定为无有效目标或颜色不在预设范围内，当前轮跳过。
- 抓取动作由 `pickObject()` 完成，内部按“到位、下探、闭合夹爪”的顺序执行。
- 放置动作只改变底座角度，不同颜色落到不同区域，最后统一回到初始姿态。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 颜色识别 | 串口能输出 RGB 数据且分类结果稳定 |
| 抓取动作 | 机械臂能完成下探和夹取流程 |
| 分类放置 | 红绿蓝目标能被放到对应区域 |
| 复位动作 | 每轮结束后机械臂回到统一原点 |


## 可扩展方向

- 增加传送带，形成连续上料分拣流程。
- 增加更多颜色阈值和异常物体处理逻辑。
- 用视觉识别替代基础颜色传感器。
- 加入动作录制与标定界面，便于不同结构快速适配。
