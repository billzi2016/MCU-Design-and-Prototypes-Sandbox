# 小型机械臂分拣系统

## 项目简介

本项目实现一个基于 Arduino、颜色传感器和四舵机机械臂的小型分拣系统，使用颜色传感器识别红、绿、蓝三类目标物体，再控制机械臂执行抓取和按颜色分类放置动作。


## 主要器件

Arduino、颜色传感器、四路舵机、机械臂结构、分拣区域底座。


## 核心功能

- 红绿蓝颜色识别。
- 机械臂抓取动作。
- 按颜色分类放置。
- 机械臂回原点。


## 运行方式

1. 打开 `src/small_robotic_arm_sorting_system/small_robotic_arm_sorting_system.ino`。
2. 连接颜色传感器和四路舵机。
3. 根据实际物体颜色调整阈值。
4. 编译上传后测试抓取和分拣流程。


## 控制逻辑说明

- 先读取 RGB 三个通道的响应。
- 判断主导颜色后决定目标分拣区。
- 机械臂执行抓取、旋转放置和复位动作。


## 验证标准

能够识别不同颜色并把物体放到对应位置，分拣结束后回到初始姿态。


## 可扩展方向

- 增加传送带。

- 增加更多颜色和类别。
