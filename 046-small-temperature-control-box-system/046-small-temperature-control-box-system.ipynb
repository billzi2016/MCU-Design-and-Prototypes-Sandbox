# 小型温控箱系统

## 项目简介

本项目实现一个基于 Arduino、DHT22、继电器和 OLED 的小型温控箱系统。系统实时采集箱内温湿度，围绕目标温度自动控制加热片和风扇，并支持本地按键修改设定温度和切换控制模式，适合作为温控实验箱或育苗箱类原型系统基础。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| DHT22 | 1 | 温湿度采集 |
| 继电器模块 | 2 | 控制加热片和风扇 |
| 加热片 / 加热丝 | 1 | 升温执行器 |
| 风扇 | 1 | 散热执行器 |
| OLED | 1 | 状态显示 |
| 按键 | 3 | 模式切换和设定值调节 |


## 核心功能

- 周期采集温湿度。
- 自动模式下根据目标温度和回差控制加热与散热。
- 支持加热、散热、空闲三种手动模式。
- 本地按键调整目标温度。
- OLED 显示当前环境、控制模式和执行器状态。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| DHT22 | D4 |
| 加热继电器 | D5 |
| 风扇继电器 | D6 |
| 模式按键 | D8 |
| 设定升高按键 | D9 |
| 设定降低按键 | D10 |
| OLED SDA / SCL | I2C 总线 |


## 接线说明

- DHT22 数据脚接 D4，加热片和风扇分别通过继电器接 D5、D6。
- 三个按键使用内部上拉输入，按下有效。
- OLED 通过 I2C 连接，用于显示环境值和控制状态。
- 加热片和风扇的电源回路应独立于主控逻辑电源设计，并与继电器隔离。


## 运行方式

1. 上传 `src/small_temperature_control_box_system/small_temperature_control_box_system.ino`。
2. 完成 DHT22、加热继电器、风扇继电器和 OLED 接线。
3. 先观察环境温湿度显示是否正常。
4. 用按键修改目标温度，验证自动模式下加热和风扇切换逻辑。
5. 逐个切换手动模式，确认执行器控制方向正确。


## 控制逻辑说明

- 自动模式下，当温度低于目标温度减回差时启动加热。
- 当温度高于目标温度加回差时启动风扇散热。
- 当温度回到目标附近时关闭执行器，避免继电器抖动。
- 若传感器故障则关闭所有执行器，优先保证安全。
- OLED 轮播显示环境信息和控制状态，兼顾运行观察与调参。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 温湿度采集 | OLED 能稳定显示实时温湿度 |
| 自动加热 | 温度低于阈值时加热继电器动作 |
| 自动散热 | 温度高于阈值时风扇继电器动作 |
| 设定调节 | 按键可修改目标温度和模式 |
