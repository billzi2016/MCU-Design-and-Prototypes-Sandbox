# 多传感器环境监测终端

## 项目简介

本项目实现一个基于 ESP32 的多传感器环境监测终端，联合使用 DHT22、MQ-2、光敏电阻和 BMP280 采集温湿度、烟雾强度、光照强度和气压信息，并通过 I2C OLED 进行分页显示。

系统不只是简单读取多个传感器，而是增加了数据平滑、舒适度判断、烟雾风险分级和总体环境状态评估，便于后续扩展为 Web Dashboard、MQTT 上传或联网节点。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| DHT22 | 1 | 温湿度采集 |
| MQ-2 | 1 | 烟雾或可燃气体强度检测 |
| 光敏电阻 | 1 | 光照强度采集 |
| BMP280 | 1 | 气压检测 |
| 0.96 寸 I2C OLED | 1 | 分页显示监测信息 |
| 10k 电阻 | 1 | 与光敏电阻组成分压 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 采集温度与湿度。
- 采集烟雾或可燃气体强度。
- 采集环境光照强度。
- 采集大气压强。
- OLED 自动轮播显示不同信息页面。
- 对模拟量传感器做平滑处理，减少抖动。
- 根据烟雾和舒适度情况给出 `NORMAL`、`WARN`、`DANGER` 状态。
- 保留联网扩展空间，适合作为后续数据上传节点。


## 引脚分配

| 功能 | ESP32 引脚 |
| --- | --- |
| DHT22 数据线 | GPIO4 |
| MQ-2 模拟输出 | GPIO34 |
| 光敏电阻分压输出 | GPIO35 |
| OLED SDA | GPIO21 |
| OLED SCL | GPIO22 |
| BMP280 SDA | GPIO21 |
| BMP280 SCL | GPIO22 |


## 接线说明

### DHT22 接线

| DHT22 | ESP32 |
| --- | --- |
| VCC | 3.3V |
| DATA | GPIO4 |
| GND | GND |

### MQ-2 接线

| MQ-2 | ESP32 |
| --- | --- |
| VCC | 5V |
| GND | GND |
| AO | GPIO34 |

### 光敏电阻接线

光敏电阻与 10k 电阻组成分压电路，分压中点接到 GPIO35。

| 元件连接 | 说明 |
| --- | --- |
| 光敏电阻一端 | 3.3V |
| 光敏电阻另一端 | GPIO35 和 10k 电阻一端 |
| 10k 电阻另一端 | GND |

### OLED 与 BMP280 接线

| 模块 | ESP32 |
| --- | --- |
| OLED SDA | GPIO21 |
| OLED SCL | GPIO22 |
| BMP280 SDA | GPIO21 |
| BMP280 SCL | GPIO22 |
| OLED/BMP280 VCC | 3.3V |
| OLED/BMP280 GND | GND |


## 运行方式

1. 使用 Arduino IDE 打开 `src/multi_sensor_environment_monitoring_terminal/multi_sensor_environment_monitoring_terminal.ino`。
2. 安装 `DHT sensor library` 库。
3. 安装 `Adafruit_BMP280` 库。
4. 安装 `Adafruit_GFX` 库。
5. 安装 `Adafruit_SSD1306` 库。
6. 开发板选择 `ESP32` 对应型号。
7. 按照接线说明完成连接。
8. 编译并上传程序。


## 页面与状态设计

- 页面 1：总览页面，显示温湿度、气压和总体状态。
- 页面 2：烟雾与光照页面，显示 MQ-2 和光照原始强度。
- 页面 3：舒适度页面，显示温度、湿度和气压状态。
- 页面 4：节点信息页面，显示系统组成和采样节奏。
- 对 MQ-2 和光敏电阻采样值做一阶低通平滑，避免页面数值抖动过快。
- 根据烟雾风险、温湿度舒适区和 BMP280 工作状态综合给出总体环境等级。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 温湿度采集 | OLED 正常显示温度和湿度数据 |
| 气压采集 | BMP280 工作正常时可显示气压值 |
| 烟雾采集 | 靠近烟雾源后 MQ-2 数值明显上升 |
| 光照采集 | 遮挡光敏电阻后光照值发生变化 |
| 页面切换 | OLED 自动轮播不同监测页面 |
| 状态评估 | 当烟雾过大或舒适度异常时，总体状态变为 `WARN` 或 `DANGER` |
| 平滑效果 | 模拟量数据不会剧烈跳变 |


## 可扩展方向
