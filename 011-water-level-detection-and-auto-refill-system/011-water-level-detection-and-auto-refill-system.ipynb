# 水位检测与自动补水系统

## 项目简介

本项目实现一个基于 Arduino Uno 的水位检测与自动补水系统，使用超声波模块检测液面距离，使用水位探头辅助判断高水位状态，通过继电器控制水泵自动补水，并使用 LCD1602 I2C 显示当前水位和运行状态。

系统加入了自动补水启动阈值、停止阈值、水泵最长运行时间和停机冷却时间，避免水泵频繁启停或长时间空转，更接近实际储水箱自动补水控制场景。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| HC-SR04 | 1 | 超声波液位测量 |
| 水位探头模块 | 1 | 高水位辅助判断 |
| 继电器模块 | 1 | 控制补水泵启停 |
| 水泵 | 1 | 执行补水 |
| 有源蜂鸣器 | 1 | 低水位或故障提示 |
| LCD1602 I2C | 1 | 显示水位和状态 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 超声波测量液面距离并换算水位百分比。
- 水位探头辅助检测高水位状态。
- 水位过低时自动启动补水。
- 水位恢复或达到高位条件时自动停止补水。
- 对补水泵设置最长运行时长和冷却时间。
- LCD1602 实时显示水位、探头值和工作状态。
- 传感器故障或临界低水位时蜂鸣报警。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| HC-SR04 Trig | D6 |
| HC-SR04 Echo | D7 |
| 水位探头模拟输出 | A0 |
| 继电器控制 | D8 |
| 蜂鸣器 | D9 |
| LCD SDA | A4 |
| LCD SCL | A5 |


## 运行方式

1. 使用 Arduino IDE 打开 `src/water_level_detection_and_auto_refill_system/water_level_detection_and_auto_refill_system.ino`。
2. 安装 `LiquidCrystal_I2C` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 水位检测 | LCD 正确显示当前水位百分比 |
| 自动补水 | 水位低于启动阈值后继电器吸合，水泵启动 |
| 停止补水 | 水位恢复或探头检测到高水位后水泵停止 |
| 运行保护 | 水泵连续运行达到上限后自动停止 |
| 故障提示 | 超声波测量异常时 LCD 显示故障信息 |
