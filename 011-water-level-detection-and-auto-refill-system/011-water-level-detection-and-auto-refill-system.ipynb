# 水位检测与自动补水系统

## 项目简介

本项目实现一个基于 Arduino Uno 的水位检测与自动补水系统，使用超声波模块测量液面距离，使用模拟水位探头辅助判断高水位，通过继电器控制水泵自动补水，并在 LCD1602 I2C 上显示当前水位、探头值和补水状态。

系统在控制逻辑中加入了补水启动阈值、停止阈值、最长运行时间和冷却等待时间，目的是避免水泵频繁启停、长时间空转或由于液面波动造成的反复触发，更接近实际水箱补水原型。

## 核心功能

- 超声波测量液面距离并换算水位百分比。
- 水位探头用于辅助判断高水位。
- 低水位自动启动补水泵。
- 水位恢复、高位探头触发或超时后自动停泵。
- 对补水泵设置冷却时间，避免频繁启停。
- LCD1602 两页轮播显示运行结果和采样细节。
- 严重低水位或传感器故障时蜂鸣报警。

## 运行方式

1. 使用 Arduino IDE 打开 `src/water_level_detection_and_auto_refill_system/water_level_detection_and_auto_refill_system.ino`。
2. 安装 `LiquidCrystal_I2C` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。
6. 根据实际水箱高度和探头阈值，调整代码中的标定参数。


## 控制逻辑说明

- 使用超声波测得液面距离，再结合安装偏移量换算成水位高度和百分比。
- 对水位百分比做一阶低通平滑，减少液面轻微波动引起的误触发。
- 当水位低于启动阈值且满足冷却时间后，系统进入补水状态。
- 当水位恢复、高水位探头触发或水泵运行超时后，系统立即停泵。
- 连续测量失败后进入传感器故障状态，并停止补水输出。
- LCD 两页轮播显示结果页和细节页，便于调试标定。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 水位显示 | LCD 正确显示当前水位百分比 |
| 自动补水 | 水位低于阈值后继电器吸合，水泵启动 |
| 自动停泵 | 水位恢复或高水位探头触发后水泵停止 |
| 超时保护 | 水泵连续运行达到上限后自动停止 |
| 故障提示 | 超声波持续异常时 LCD 显示故障信息并停止补水 |
| 冷却保护 | 刚停泵后不会立刻再次启动水泵 |


## 可扩展方向

- 增加流量计，实现按补水量闭环控制。
- 增加 GSM 或 WiFi 模块，实现远程报警。
- 增加 EEPROM 保存阈值与标定参数。
- 增加多水箱切换控制，扩展为多路补水系统。
- 增加历史记录，用于观察补水频率和液位变化。
