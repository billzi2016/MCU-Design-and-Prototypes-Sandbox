# 步进电机精确控制平台

## 项目简介

本项目实现一个基于 Arduino Uno、A4988 或 DRV8825 的步进电机精确控制平台，支持上电回零、角度定位、速度切换和串口命令控制。通过限位开关建立零点参考，适合作为滑台、转台或自动定位机构的基础控制模块。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| A4988/DRV8825 | 1 | 步进驱动器 |
| 步进电机 | 1 | 执行机构 |
| 限位开关 | 1 | 回零参考 |
| 按键 | 3 | 本地控制 |


## 核心功能

- 限位回零。
- 角度控制。
- 速度切换。
- 串口指令控制。
- 回零超时保护。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| STEP | D2 |
| DIR | D3 |
| ENABLE | D4 |
| HOME 限位 | D5 |
| HOME 按键 | D6 |
| 目标角度按键 | D7 |
| 速度按键 | D8 |


## 接线说明

步进驱动器的 `STEP`、`DIR`、`ENABLE` 分别接到 D2、D3、D4；限位开关一端接 D5，另一端接 GND；按键均使用 `INPUT_PULLUP`，一端接对应引脚，另一端接 GND。


## 运行方式

1. 打开 `src/precise_stepper_motor_control_platform/precise_stepper_motor_control_platform.ino`。
2. 安装 `AccelStepper` 库。
3. 完成步进驱动器、限位开关和按键接线。
4. 编译上传后测试回零和角度控制。


## 控制逻辑说明

- 上电先执行回零，建立位置参考。
- 目标角度统一换算为细分步数。
- 本地按钮与串口命令共用同一运动接口。
- 回零超时后不再继续运动，避免限位失效时死转。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 回零动作 | 上电后能移动到限位并置零 |
| 角度定位 | 切换目标角度后步进电机到达对应位置 |
| 速度切换 | 不同档位下运动速度明显变化 |
| 串口控制 | `HOME`、`ANGLE`、`SPEED` 指令可生效 |


## 可扩展方向

- 增加编码器做闭环位置校验。
- 增加多轴联动控制。
- 接入 OLED 做本地参数显示。
- 扩展为滑台或转台平台。
