# 超声波避障小车

## 项目简介

本项目实现一个基于 Arduino、HC-SR04、舵机和双电机驱动的超声波避障小车。系统先测量正前方距离，若发现障碍，再驱动舵机将超声波模块转向左右两侧进行二次扫描，比较可通行空间后选择转向方向，实现基础自主避障。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino 主控板 | 1 | 执行测距与决策 |
| HC-SR04 | 1 | 测量障碍物距离 |
| 舵机 | 1 | 带动测距模块左右扫描 |
| 双路电机驱动 | 1 | 驱动左右轮 |
| 直流减速电机 | 2 | 小车动力输出 |


## 核心功能

- 正前方实时测距。
- 障碍出现时自动左右扫描。
- 比较左右距离后选择更安全方向。
- 支持前进、后退、原地转向和停车。
- 超声波超时返回容错值，避免误判导致卡死。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| HC-SR04 Trig | D12 |
| HC-SR04 Echo | D13 |
| 扫描舵机 | D3 |
| 左电机 PWM / 方向 | D5、D6、D7 |
| 右电机 PWM / 方向 | D9、D10、D11 |


## 接线说明

- HC-SR04 的 `Trig` 接 D12，`Echo` 接 D13，并确保与主控共地。
- 舵机信号线接 D3，电源建议使用独立稳压电源，避免舵机动作导致主控掉压重启。
- 左右电机通过电机驱动接入主控，PWM 控速，方向脚负责前后切换。
- 超声波模块安装在舵机云台上时，要保证左右扫描不会被车体结构遮挡。


## 运行方式

1. 打开 `src/ultrasonic_obstacle_avoidance_car/ultrasonic_obstacle_avoidance_car.ino`。
2. 完成 HC-SR04、舵机和电机驱动接线。
3. 上传程序后先观察舵机是否能回到中位 90 度。
4. 在地面放置障碍物，测试前进、扫描和绕行流程。
5. 依据底盘尺寸调整 `SAFE_DISTANCE_CM` 和转向时间。


## 控制逻辑说明

- 主循环优先扫描正前方，若距离大于安全阈值则保持前进。
- 当正前方过近时先停车，等待惯性消除后再依次测左、测右。
- 若左侧更远且大于安全距离，则左转；若右侧更远则右转。
- 若左右都不安全，则先短暂后退脱离死角，再执行固定方向转向。
- `readDistanceCm()` 超时返回较大距离值，防止单次测距失败把系统误锁在障碍状态。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 前方无障碍 | 小车能持续前进 |
| 前方有障碍 | 小车会停车并启动左右扫描 |
| 空间比较 | 左右距离不同时时能选择更大空间方向 |
| 死角处理 | 左右都近时会后退后再转向 |


## 可扩展方向

- 增加蓝牙模式，支持手动 / 自动切换。
- 增加红外、碰撞开关等多传感器融合避障。
- 增加路径记忆与简单地图构建。
- 将固定转向时长升级为基于编码器的角度控制。
