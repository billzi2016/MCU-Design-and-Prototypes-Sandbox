# 超声波避障小车

## 项目简介

本项目实现一个基于 Arduino、HC-SR04、舵机和双电机驱动的超声波避障小车，通过舵机带动超声波模块扫描左右和前方距离，根据最优方向自动转向，实现基础自主避障。


## 主要器件

Arduino、小车底盘、HC-SR04、舵机、双电机驱动、直流减速电机。


## 核心功能

- 前方距离测量。
- 左右方向扫描。
- 自动决策转向。
- 前进、后退、转向控制。


## 运行方式

1. 打开 `src/ultrasonic_obstacle_avoidance_car/ultrasonic_obstacle_avoidance_car.ino`。
2. 连接 HC-SR04、舵机和双电机驱动。
3. 编译上传后在障碍物环境中测试。
4. 视底盘情况调整安全距离和转向时长。


## 控制逻辑说明

- 先测前方距离。
- 若前方安全则继续前进。
- 若前方不安全则扫描左右距离。
- 选择空间更大的方向转向，若两侧都不安全则后退再转向。


## 验证标准

前方有障碍时能停车并扫描，最终绕开障碍继续运动。


## 可扩展方向

- 增加蓝牙手动 / 自动切换。
- 多传感器避障。
- 地图构建。
