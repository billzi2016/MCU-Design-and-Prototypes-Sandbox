# 12864 图形液晶菜单显示系统

## 项目简介

本项目实现一个基于 Arduino Uno 和 12864 图形液晶的菜单显示与参数设置系统。系统通过按键完成菜单切换、参数调整和确认操作，并在 12864 屏幕上显示当前页面、参数值和运行状态。

12864 液晶相比 1602 字符屏可以显示更多信息，适合用于传感器终端、控制器面板和小型仪器的人机交互界面。Arduino Uno 可以满足按键输入和基础图形显示需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 12864 图形液晶模块 | 1 | 菜单和参数显示 |
| 轻触按键 | 4 | 上、下、确认、返回 |
| 10k 电位器 | 1 | 液晶对比度调节，部分模块需要 |
| 220Ω 电阻 | 1 | 背光限流，按模块情况使用 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 12864 屏幕显示系统标题、菜单项和状态信息。
- 支持上、下按键切换菜单项。
- 支持确认键进入参数设置。
- 支持返回键退出设置页面。
- 支持亮度阈值、报警阈值和采样间隔三个参数调整。
- 当前选中项使用箭头标识。
- 按键使用软件消抖，避免一次按下触发多次。


## 引脚分配

本项目使用 ST7920 控制器的 12864 液晶，并采用串行连接方式，减少 Arduino Uno 引脚占用。

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 12864 E / SCK | D13 |
| 12864 R/W / SID | D11 |
| 12864 RS / CS | D10 |
| 上按键 | D2 |
| 下按键 | D3 |
| 确认按键 | D4 |
| 返回按键 | D5 |


## 接线说明

### 12864 液晶接线

| 12864 模块 | Arduino Uno |
| --- | --- |
| VSS | GND |
| VDD | 5V |
| V0 | 电位器中间脚，用于对比度调节 |
| RS / CS | D10 |
| R/W / SID | D11 |
| E / SCK | D13 |
| PSB | GND，选择串行模式 |
| RST | 5V |
| BLA | 5V，按模块情况可串联限流电阻 |
| BLK | GND |

不同 12864 模块的引脚命名可能略有差异，接线前应核对模块丝印。代码按 ST7920 串行模式编写。

### 按键接线

程序使用 Arduino Uno 内部上拉电阻，因此每个按键一端接对应数字引脚，另一端接 GND。按键未按下时，引脚读数为 HIGH；按键按下时，引脚被拉低为 LOW。

| 按键功能 | 一端连接 | 另一端连接 |
| --- | --- | --- |
| 上 | D2 | GND |
| 下 | D3 | GND |
| 确认 | D4 | GND |
| 返回 | D5 | GND |


## 运行方式

1. 使用 Arduino IDE 打开 `src/menu_12864/menu_12864.ino`。
2. 安装 `U8g2` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。


## 状态变量

- `selectedIndex`：当前选中的菜单项。
- `editMode`：是否处于参数编辑模式。
- `brightnessThreshold`：亮度阈值。
- `alarmThreshold`：报警阈值。
- `sampleInterval`：采样间隔。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |

| 参数编辑 | 确认键可以进入编辑模式 |
| 参数调整 | 编辑模式下上、下按键可以修改参数 |

## 可扩展方向

- 增加多级菜单。
