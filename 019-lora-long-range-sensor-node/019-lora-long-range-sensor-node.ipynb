# LoRa 远距离传感器节点

## 项目简介

本项目实现一个基于 ESP32、DHT22 和 LoRa 模块的远距离传感器节点，周期性采集温湿度并通过 LoRa 广播发送数据，同时用 OLED 显示节点编号、传感器状态和已发送的数据包数量。

该实现偏向发射端节点，适合作为农业、仓储或野外环境监测系统中的远距离采集端。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| LoRa 模块 | 1 | 远距离无线通信 |
| DHT22 | 1 | 温湿度检测 |
| 0.96 寸 I2C OLED | 1 | 本地状态显示 |
| 天线 | 1 | LoRa 发射天线 |


## 核心功能

- 温湿度采集。
- LoRa 周期广播发送。
- 节点编号标识。
- OLED 显示无线状态和发包数量。
- 传感器无效时停止发包。


## 引脚分配

| 功能 | ESP32 引脚 |
| --- | --- |
| DHT22 数据线 | GPIO4 |
| LoRa SS | GPIO5 |
| LoRa RST | GPIO14 |
| LoRa DIO0 | GPIO2 |
| OLED SDA | GPIO21 |
| OLED SCL | GPIO22 |


## 接线说明

| 模块 | ESP32 |
| --- | --- |
| DHT22 DATA | GPIO4 |
| LoRa NSS/CS | GPIO5 |
| LoRa RST | GPIO14 |
| LoRa DIO0 | GPIO2 |
| OLED SDA | GPIO21 |
| OLED SCL | GPIO22 |

LoRa 模块频率需与网关端一致，天线必须正确连接后再上电发射。


## 运行方式

1. 使用 Arduino IDE 打开 `src/lora_long_range_sensor_node/lora_long_range_sensor_node.ino`。
2. 安装 `LoRa`、`DHT sensor library`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 根据实际模块频率调整 `LORA_FREQUENCY`。
4. 编译并上传程序。
5. 在接收端或网关侧确认数据包是否成功收到。


## 控制逻辑说明

- 节点按固定周期读取温湿度数据。
- 传感器有效时，构造包含节点编号、温度、湿度和包计数的 JSON 字符串。
- LoRa 发射采用广播方式，不依赖点对点确认流程，适合基础节点原型。
- OLED 本地显示节点编号、无线状态、传感器状态和累计发包数。

## 可扩展方向

- 增加低功耗休眠与电池供电策略。
- 增加多种环境传感器。
