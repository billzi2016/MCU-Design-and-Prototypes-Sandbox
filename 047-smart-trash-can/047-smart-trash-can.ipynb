# 智能垃圾桶

## 项目简介

本项目实现一个基于 Arduino、红外接近检测、超声波液位检测、舵机和蜂鸣器的智能垃圾桶。系统能够在用户靠近时自动开盖，并通过超声波检测桶内剩余空间，在接近满载时发出蜂鸣提示，适合作为智能家居类演示原型。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 红外接近传感器 | 1 | 检测用户靠近 |
| HC-SR04 | 1 | 检测桶内剩余高度 |
| 舵机 | 1 | 控制桶盖开合 |
| 蜂鸣器 | 1 | 满载提示 |
| LCD1602 I2C | 1 | 状态显示 |


## 核心功能

- 用户靠近时自动开盖。
- 保持开盖一段时间后自动关闭。
- 超声波检测满载程度。
- 满载时蜂鸣报警。
- LCD 显示桶盖状态和剩余空间。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 红外接近传感器 | D2 |
| 舵机 | D9 |
| 蜂鸣器 | D8 |
| HC-SR04 Trig / Echo | D12、D13 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- 红外接近传感器安装在桶盖前方，用于检测手部靠近。
- 超声波模块安装在桶盖上方向桶内测距，距离越小说明越接近满载。
- 舵机连接桶盖机构，开闭角度需按实际结构重新校准。
- 蜂鸣器用于满载提示，LCD 用于显示实时状态。


## 运行方式

1. 上传 `src/smart_trash_can/smart_trash_can.ino`。
2. 完成红外、超声波、舵机、蜂鸣器和 LCD 接线。
3. 测试靠近时桶盖是否自动打开。
4. 调整满载阈值，验证垃圾接近桶口时蜂鸣报警。
5. 调整保持开盖时间和舵机角度以适配机械结构。


## 控制逻辑说明

- 红外传感器负责快速触发开盖动作。
- 桶盖打开后保持一段时间，再自动关闭。
- 超声波周期检测桶内剩余高度，低于阈值视为满载。
- 满载时继续允许显示提示，但不再反复自动开盖，避免溢出风险。
- LCD 用于同时反馈桶盖状态和当前满载距离值。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 自动开盖 | 手靠近时舵机能打开桶盖 |
| 自动闭盖 | 保持时间结束后桶盖关闭 |
| 满载检测 | 垃圾接近阈值时报警触发 |
| 状态显示 | LCD 能显示桶盖状态和空间信息 |


## 可扩展方向

- 增加重量检测形成更准确的满载判断。
- 接入 WiFi 上报垃圾桶状态。
- 增加分类投放识别。
- 增加电池供电与低功耗唤醒设计。
