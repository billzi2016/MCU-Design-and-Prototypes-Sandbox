# RFID 门禁与考勤系统

## 项目简介

本项目实现一个基于 Arduino Uno、RC522、LCD1602 I2C 和舵机的 RFID 门禁与考勤系统，识别授权卡后自动开门，并在内存中累加该用户的考勤次数；未授权卡会被拒绝并提示错误。

该项目把门禁控制与考勤统计合并在同一套原型中，适合作为宿舍门禁、实验室出入管理或课堂签到的基础演示系统。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| RC522 | 1 | RFID 读卡模块 |
| RFID 卡片 | 若干 | 门禁卡或考勤卡 |
| LCD1602 I2C | 1 | 显示刷卡结果 |
| SG90 舵机 | 1 | 模拟开门 |
| 有源蜂鸣器 | 1 | 成功或失败提示 |


## 核心功能

- 刷卡识别。
- 权限判断。
- 授权卡自动开门。
- 考勤次数统计。
- LCD 显示欢迎信息或拒绝信息。
- 蜂鸣器区分成功和失败提示。

| 模块 | Arduino Uno |
| --- | --- |
| RC522 SS | D10 |
| RC522 RST | D9 |
| RC522 SPI 其余引脚 | 对应默认 SPI 引脚 |
| 舵机信号 | D6 |
| 蜂鸣器正极 | D8 |
| 蜂鸣器负极 | GND |

舵机建议使用独立 5V 供电，外部电源地与 Arduino GND 共地。


## 运行方式

1. 使用 Arduino IDE 打开 `src/rfid_access_control_and_attendance_system/rfid_access_control_and_attendance_system.ino`。
2. 安装 `MFRC522`、`LiquidCrystal_I2C` 和 `Servo` 库。
3. 根据实际卡 UID 修改 `authorizedCards` 数组。
4. 编译并上传程序。
5. 依次测试授权卡和未授权卡的刷卡结果。


## 控制逻辑说明

- RC522 检测到新卡后读取 UID，并与授权列表做匹配。
- 授权成功时打开舵机门锁，累加考勤次数，并显示欢迎信息。
- 未授权时拒绝开门，同时把 UID 以字符串形式显示出来，便于识别未知卡。
- 门锁打开只维持固定时间，超时后自动关闭。
- LCD 优先显示临时消息，消息结束后恢复到待刷卡主页。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 授权卡识别 | 授权卡刷卡后显示欢迎信息并开门 |
| 未授权卡拒绝 | 非授权卡刷卡后拒绝访问 |
| 考勤统计 | 同一授权卡多次刷卡后计数递增 |
| 自动上锁 | 到达设定时长后门锁自动关闭 |
| 蜂鸣提示 | 成功与失败的蜂鸣节奏不同 |


## 可扩展方向

- 增加 EEPROM 保存考勤计数。
- 增加管理员卡和卡片注册功能。
- 增加 SD 卡或 WiFi 上传考勤日志。
- 将舵机替换为电磁锁或继电器控制门锁。
- 与人脸识别或指纹模块融合成多因子门禁系统。
