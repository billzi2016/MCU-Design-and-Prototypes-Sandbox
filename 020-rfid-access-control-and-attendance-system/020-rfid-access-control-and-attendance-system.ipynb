# RFID 门禁与考勤系统

## 项目简介

本项目实现一个基于 Arduino Uno、RC522、LCD1602 I2C 和舵机的 RFID 门禁与考勤系统，识别授权卡后自动开门，并在内存中累加该用户的考勤次数；未授权卡会被拒绝并提示错误。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| RC522 | 1 | RFID 读卡模块 |
| RFID 卡片 | 若干 | 门禁卡或考勤卡 |
| LCD1602 I2C | 1 | 显示刷卡结果 |
| SG90 舵机 | 1 | 模拟开门 |
| 有源蜂鸣器 | 1 | 成功或失败提示 |


## 核心功能

- 刷卡识别。
- 权限判断。
- 授权卡自动开门。
- 考勤次数统计。
- LCD 显示欢迎信息或拒绝信息。
