# 室内空气质量监测系统

## 项目简介

本项目实现一个基于 ESP32 的室内空气质量监测系统，使用 MQ-135 检测空气质量，使用 DHT22 采集温湿度，使用 OLED 显示实时状态，并使用蜂鸣器在空气质量危险或传感器故障时发出警报。

系统将空气状态划分为 `NORMAL`、`WARNING`、`DANGER` 和 `FAULT` 四类，适合作为宿舍、办公室或实验室的本地监测终端。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| MQ-135 | 1 | 空气质量检测 |
| DHT22 | 1 | 温湿度检测 |
| 0.96 寸 I2C OLED | 1 | 显示监测结果 |
| 有源蜂鸣器 | 1 | 危险报警 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 采集空气质量模拟值。
- 采集温度和湿度。
- OLED 多页面显示空气状态和阈值信息。
- 空气质量危险时蜂鸣器报警。
- DHT22 异常时进入故障状态。


## 引脚分配

| 功能 | ESP32 引脚 |
| --- | --- |
| DHT22 数据线 | GPIO4 |
| MQ-135 模拟输出 | GPIO34 |
| 蜂鸣器 | GPIO18 |
| OLED SDA | GPIO21 |
| OLED SCL | GPIO22 |


## 运行方式

1. 使用 Arduino IDE 打开 `src/indoor_air_quality_monitoring_system/indoor_air_quality_monitoring_system.ino`。
2. 安装 `DHT sensor library`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 开发板选择对应 `ESP32` 型号。
4. 完成接线后编译并上传程序。
