# 室内空气质量监测系统

## 项目简介

本项目实现一个基于 ESP32 的室内空气质量监测系统，使用 MQ-135 检测空气质量，使用 DHT22 采集温湿度，使用 OLED 显示空气状态、舒适度和阈值信息，并在空气质量危险或传感器故障时通过蜂鸣器发出报警。

系统把状态划分为 `NORMAL`、`WARNING`、`DANGER` 和 `FAULT` 四类，更适合做宿舍、办公室或实验室的本地监测终端，也便于后续继续扩展联网和历史记录功能。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| MQ-135 | 1 | 空气质量检测 |
| DHT22 | 1 | 温湿度检测 |
| 0.96 寸 I2C OLED | 1 | 状态显示 |
| 有源蜂鸣器 | 1 | 危险报警 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 采集空气质量模拟值。
- 采集温度和湿度。
- 按空气质量和舒适度划分状态等级。
- OLED 轮播显示总览、阈值和舒适度信息。
- 空气危险或传感器故障时蜂鸣器报警。
- 对模拟量采样做平滑处理，减少数值抖动。

## 运行方式

1. 使用 Arduino IDE 打开 `src/indoor_air_quality_monitoring_system/indoor_air_quality_monitoring_system.ino`。
2. 安装 `DHT sensor library`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
3. 开发板选择对应 `ESP32` 型号。
4. 完成接线后编译并上传程序。
5. 根据实际模块输出范围调整空气质量阈值。


## 控制逻辑说明

- 每 2 秒读取一次 DHT22 和 MQ-135 数据。
- DHT22 读数失败时进入故障状态，并在屏幕中提示传感器异常。
- 对 MQ-135 模拟值做一阶低通平滑，再用于状态判断和显示。
- 当空气质量超过高阈值时进入 `DANGER`；超过普通阈值或环境闷热潮湿时进入 `WARNING`。
- 蜂鸣器只在 `DANGER` 或 `FAULT` 状态下工作，避免一般预警过于扰民。
- OLED 自动轮播总览页、阈值页和舒适度页。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 空气质量采集 | OLED 显示 MQ-135 数值 |
| 温湿度采集 | OLED 正确显示温湿度 |
| 页面轮播 | 3 个页面能自动切换 |
| 预警状态 | 空气质量偏高或环境过热潮湿时进入 `WARNING` |
| 危险状态 | 空气质量明显过高时进入 `DANGER` 并蜂鸣报警 |
| 故障状态 | DHT22 异常时进入 `FAULT` |


## 可扩展方向

- 增加 WiFi 上传，接入网页或手机端监测。
- 增加继电器，联动排风扇或空气净化器。
- 增加历史缓存，观察空气变化趋势。
- 增加按键，支持本地阈值调整。
- 替换为更多气体传感器，扩展为多指标空气质量终端。
