# WiFi 数据上传监测节点

## 项目简介

本项目实现一个基于 ESP32 的 WiFi 数据上传监测节点，使用 DHT22 采集温湿度，连接本地 WiFi，并定时向 HTTP 服务上传 JSON 数据，同时用 OLED 显示连接状态、上传结果和当前测量值。

该节点适合作为云端监测系统、局域网网页看板或实验室数据采集终端的基础版本，重点覆盖传感器采集、联网、HTTP 上传和本地状态反馈这四个环节。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| ESP32 开发板 | 1 | 主控板 |
| DHT22 | 1 | 温湿度检测 |
| 0.96 寸 I2C OLED | 1 | 状态显示 |
| WiFi 网络 | 1 | 数据上传通道 |


## 核心功能

- 温湿度采集。
- WiFi 自动连接与定时重连。
- HTTP JSON 数据上传。
- OLED 显示连接状态和上传结果。
- 记录最近一次 HTTP 状态码。

## 运行方式

1. 使用 Arduino IDE 打开 `src/wifi_data_upload_monitoring_node/wifi_data_upload_monitoring_node.ino`。
2. 修改 `WIFI_SSID`、`WIFI_PASSWORD` 和 `UPLOAD_URL`。
3. 安装 `DHT sensor library`、`Adafruit_GFX` 和 `Adafruit_SSD1306` 库。
4. 编译并上传程序。
5. 在服务器端确认是否收到 JSON 数据。


## 控制逻辑说明

- 先保证 WiFi 连通，再执行采样和上传逻辑。
- DHT22 无效时不执行上传，避免把无意义数据发到服务端。
- 到达上传间隔后，构造 JSON 数据并用 HTTP POST 发送。
- 若 WiFi 断开，系统按固定时间重连，而不是每圈都强制重连。
- OLED 显示当前 WiFi 状态、传感器值和最近一次 HTTP 状态码。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| WiFi 连接 | 上电后 ESP32 成功连接到指定网络 |
| 传感器采样 | OLED 正常显示温湿度 |
| 数据上传 | 服务端收到 POST 的 JSON 数据 |
| 断网重连 | WiFi 中断后系统能自动重连 |
| 状态显示 | OLED 能显示 HTTP 结果和连接状态 |


## 可扩展方向

- 改为 MQTT 上传。
- 增加多传感器并同时上传。
- 增加本地缓存，网络恢复后补传。
- 增加 Web 配网页面，用于在线修改参数。
- 增加阈值报警和远程推送通知。
