# 倒立摆控制系统

## 项目简介

本项目实现一个基于 Arduino 的倒立摆控制系统原型，使用角度传感输入模拟摆杆角度，通过 PID 计算小车或滑台的驱动输出，使摆杆尽量保持在竖直附近。该实现适合作为倒立摆控制课设的基础软件框架。


## 主要器件

主控板、角度传感输入、电机驱动、直流电机或滑台执行机构。


## 核心功能

- 倒立摆角度采样。
- PID 闭环控制。
- 电机输出调节。
- 超限安全停机。


## 运行方式

1. 打开 `src/inverted_pendulum_control_system/inverted_pendulum_control_system.ino`。
2. 安装 `PID_v1` 库。
3. 连接角度输入和电机驱动。
4. 编译上传后观察角度和 PWM 调试输出。


## 控制逻辑说明

- 读取角度输入并换算成角度值。
- 用 PID 计算控制量。
- 输出到滑台或小车执行机构。
- 超出安全角度时立即停机。


## 验证标准

角度变化能够引起 PWM 变化，超限后停机，串口输出连续稳定。
