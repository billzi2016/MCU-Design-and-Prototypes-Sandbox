# 倒立摆控制系统

## 项目简介

本项目实现一个基于 Arduino 的倒立摆控制系统软件原型，使用模拟角度输入代表摆杆偏角，通过 PID 计算驱动小车或滑台执行机构的输出，使摆杆尽量保持在竖直附近。该版本重点展示倒立摆控制链路和安全停机逻辑，适合作为课设控制框架基础。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| 主控板 | 1 | 执行 PID 运算 |
| 电位器或角度传感输入 | 1 | 模拟摆杆角度 |
| 电机驱动模块 | 1 | 驱动底座或滑台 |
| 电机 / 滑台执行机构 | 1 套 | 执行控制动作 |


## 核心功能

- 读取摆杆角度输入并换算为角度值。
- 使用 `PID_v1` 库进行闭环控制。
- 根据 PWM 正负控制底座运动方向。
- 摆角超出安全范围时立即停机。
- 通过串口输出角度和控制量，便于整定。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 角度输入 | A0 |
| 电机 PWM | D5 |
| 电机方向 | D6、D7 |


## 接线说明

- 若使用电位器模拟摆角，电位器两端接 5V 和 GND，中间滑动端接 A0。
- 电机驱动的方向引脚接 D6、D7，PWM 调速引脚接 D5。
- 执行机构电源建议独立供给，并与主控共地，防止驱动噪声影响模拟输入。
- 如果后续换成真实角度传感器，只需要将 `readAngleDegrees()` 换成对应的读取函数。


## 运行方式

1. 打开 `src/inverted_pendulum_control_system/inverted_pendulum_control_system.ino`。
2. 安装 `PID_v1` 库。
3. 连接角度输入与电机驱动模块。
4. 上传程序后观察串口中的角度与 PWM 输出。
5. 逐步调节 `kp`、`ki`、`kd` 以改善响应。


## 控制逻辑说明

- 主循环使用固定控制周期，避免控制节拍抖动影响 PID 效果。
- 先采集当前摆角，再与 `setpointAngle` 比较生成误差。
- `PID_v1` 根据误差输出正负控制量，正值和负值分别代表底座向不同方向运动。
- 当摆角超过 `SAFE_ANGLE_DEG` 时直接停机，防止底座猛烈冲撞行程端点。
- 当前版本使用电位器模拟角度输入，适合先验证软件控制链路。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 角度读取 | 转动电位器时串口角度连续变化 |
| PWM 输出 | 角度偏向两侧时输出方向会切换 |
| 超限停机 | 超过安全角度后电机停止 |
| PID 整定 | 修改参数后控制反应明显变化 |


## 可扩展方向

- 接入真实编码器和角度传感器构成完整倒立摆平台。
- 增加位置环，实现防止底座越走越远的串级控制。
- 与 LQR、状态空间控制器做效果对比。
- 增加上位机曲线监视和参数在线调整。
