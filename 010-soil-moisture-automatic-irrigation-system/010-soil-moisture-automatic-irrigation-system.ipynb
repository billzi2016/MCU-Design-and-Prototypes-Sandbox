# 土壤湿度自动灌溉系统

## 项目简介

本项目实现一个基于 Arduino Uno 的土壤湿度自动灌溉系统，使用土壤湿度传感器采集盆栽或土壤含水情况，使用继电器驱动水泵完成自动浇水，并通过 LCD1602 I2C 显示湿度、阈值和水泵状态。

系统加入了阈值设置、自动模式开关、手动浇水、泵最大运行时长和冷却等待机制，避免继电器频繁抖动和水泵长时间空转，更接近实际可用的灌溉控制逻辑。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 土壤湿度传感器 | 1 | 检测土壤湿度 |
| 微型水泵 | 1 | 执行浇水 |
| 继电器模块 | 1 | 控制水泵启停 |
| LCD1602 I2C | 1 | 显示湿度、阈值和工作状态 |
| 轻触按键 | 3 | 阈值设置、模式切换、手动浇水 |
| 外部电源 | 1 | 为水泵供电 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 实时检测土壤湿度。
- 按阈值自动启动灌溉。
- 支持手动启动浇水。
- 支持在线调整湿度阈值。
- LCD1602 实时显示湿度、阈值和水泵状态。
- 自动模式可开关切换。
- 设定泵最大运行时长，防止连续运行过久。
- 设定泵停机冷却时间，避免频繁启停。

## 接线说明

### 土壤湿度传感器接线

| 传感器 | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| AO | A0 |

### 继电器与水泵接线

| 模块 | 连接 |
| --- | --- |
| 继电器 IN | D8 |
| 继电器 VCC | 5V |
| 继电器 GND | GND |
| 水泵电源回路 | 串接到继电器常开触点 |

水泵工作电流通常较大，建议由外部电源供电，不要直接由 Arduino 5V 引脚驱动水泵。外部电源地线应与 Arduino GND 共地。

### 按键与 LCD 接线

| 模块 | Arduino Uno |
| --- | --- |
| 设置按键 | D2 到 GND |
| 增加按键 | D3 到 GND |
| 减少按键 | D4 到 GND |
| LCD SDA | A4 |
| LCD SCL | A5 |

程序使用内部上拉电阻，按键按下时读数为 LOW。


## 运行方式

1. 使用 Arduino IDE 打开 `src/soil_moisture_automatic_irrigation_system/soil_moisture_automatic_irrigation_system.ino`。
2. 安装 `LiquidCrystal_I2C` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。


## 操作说明

- `SET` 按键：进入或退出阈值设置模式。
- `UP` 按键：设置模式下提高阈值；普通模式下切换自动模式开关。
- `DOWN` 按键：设置模式下降低阈值；普通模式下启动手动浇水，再按一次可提前停止。
- 自动模式下，当土壤湿度低于阈值一定幅度后会自动启动水泵。
- 当湿度恢复、运行超时或用户关闭自动模式时，自动浇水会停止。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 湿度显示 | LCD 正确显示当前湿度百分比 |
| 自动浇水 | 土壤湿度低于阈值后继电器吸合，水泵启动 |
| 自动停止 | 湿度恢复或运行达到上限后水泵停止 |
| 阈值设置 | 进入设置模式后可修改阈值 |
| 自动模式切换 | 可通过按键关闭或开启自动灌溉 |
| 手动浇水 | 普通模式下可手动触发浇水 |
| 冷却保护 | 刚停止浇水后不会立刻再次启动水泵 |


## 可扩展方向

- 增加 EEPROM 保存阈值，断电后仍可保留设置。
- 增加多路湿度采集，实现多盆栽分区灌溉。
- 增加 RTC，实现定时灌溉策略。
- 增加 WiFi 模块，远程查看湿度和控制灌溉。
- 联动天气数据，避免雨天重复灌溉。
