# 篮球计分器

## 项目简介

这是一个基于 Arduino Uno 的篮球计分器原型系统，用于完成比赛比分显示、加分操作、节次切换、倒计时控制和结束蜂鸣提示。

系统使用按键作为输入，LCD1602 I2C 作为显示模块，蜂鸣器作为提示输出。Arduino Uno 可以满足本项目的 I/O、计时和显示需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| LCD1602 I2C | 1 | 显示比分、节次和倒计时 |
| 轻触按键 | 9 | 加分、节次、计时、复位 |
| 有源蜂鸣器 | 1 | 倒计时结束提示 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- A 队加 1 分、2 分、3 分。
- B 队加 1 分、2 分、3 分。
- 显示当前节次。
- 支持节次切换。
- 支持比赛倒计时启动和暂停。
- 支持一键复位比分、节次和倒计时。
- 倒计时结束后蜂鸣器提示。
- 按键使用软件消抖，避免一次按下触发多次。

## 接线说明

### LCD1602 I2C

| LCD1602 I2C | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| SDA | A4 |
| SCL | A5 |

### 按键

程序使用 Arduino Uno 内部上拉电阻，因此每个按键一端接对应数字引脚，另一端接 GND。按键未按下时，引脚读数为 HIGH；按键按下时，引脚被拉低为 LOW。

| 按键功能 | 一端连接 | 另一端连接 |
| --- | --- | --- |
| A 队 +1 | D2 | GND |
| A 队 +2 | D3 | GND |
| A 队 +3 | D4 | GND |
| B 队 +1 | D5 | GND |
| B 队 +2 | D6 | GND |
| B 队 +3 | D7 | GND |
| 节次切换 | D8 | GND |
| 计时开始/暂停 | D9 | GND |
| 复位 | D10 | GND |

### 蜂鸣器

| 蜂鸣器 | Arduino Uno |
| --- | --- |
| 正极 | D11 |
| 负极 | GND |

当前代码按有源蜂鸣器处理。如果使用无源蜂鸣器，可以后续改为 `tone()` 输出。


## 运行方式

1. 使用 Arduino IDE 打开 `src/basketball_scoreboard/basketball_scoreboard.ino`。
2. 安装 `LiquidCrystal_I2C` 库。
3. 开发板选择 `Arduino Uno`。
4. 按照接线说明完成连接。
5. 编译并上传程序。


## 状态变量

- `scoreA`：A 队比分。
- `scoreB`：B 队比分。
- `period`：当前节次。
- `remainingSeconds`：剩余秒数。
- `timerRunning`：计时器是否运行。


## 验证标准
