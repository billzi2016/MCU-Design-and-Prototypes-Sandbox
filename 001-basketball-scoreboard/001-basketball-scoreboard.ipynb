# 篮球计分器

## 设计目标

本项目设计一个基于 Arduino Uno 的篮球计分器，实现两队比分记录、节次显示、比赛倒计时、开始暂停控制和结束蜂鸣提示。


## 平台选择

- 主控平台：Arduino Uno
- 选择理由：本项目只需要按键输入、LCD 显示、蜂鸣器输出和毫秒级计时，Arduino Uno 的引脚数量和性能足够。


## 输入输出设计

| 类型 | 内容 |
| --- | --- |
| 输入 | A 队加分按键、B 队加分按键、节次按键、计时按键、复位按键 |
| 处理 | 按键消抖、比分更新、节次更新、倒计时更新 |
| 输出 | LCD 显示比分和时间、蜂鸣器提示 |


## 状态变量

- `scoreA`：A 队比分。
- `scoreB`：B 队比分。
- `period`：当前节次。
- `remainingSeconds`：剩余秒数。
- `timerRunning`：计时器是否运行。
