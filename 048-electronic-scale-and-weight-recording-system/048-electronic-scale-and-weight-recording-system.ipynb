# 电子秤与称重记录系统

## 项目简介

本项目实现一个基于 Arduino、HX711、称重传感器和 LCD 的电子秤与称重记录系统。系统支持重量采集、去皮、校准系数调整和称重记录计数，并能将记录值通过串口输出，适合作为小型称重终端或计量实验原型。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| HX711 | 1 | 称重放大与转换 |
| 称重传感器 | 1 | 采集重量 |
| LCD1602 I2C | 1 | 重量显示 |
| 按键 | 4 | 去皮、记录、校准加减 |


## 核心功能

- 多次平均采样提升称重稳定性。
- 去皮操作支持空盘扣重。
- 校准系数可现场微调。
- 记录按钮触发称重日志计数和串口输出。
- LCD 同时显示当前重量、校准系数和记录次数。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| HX711 DT / SCK | D3、D2 |
| 去皮按键 | D8 |
| 记录按键 | D9 |
| 校准增加 / 减少 | D10、D11 |
| LCD SDA / SCL | I2C 总线 |


## 接线说明

- 称重传感器接入 HX711，再由 HX711 连接 Arduino 的 DT 和 SCK。
- 四个按键均使用内部上拉输入，按下有效。
- LCD1602 通过 I2C 接口连接。
- 首次使用时需要结合标准砝码调节 `calibrationFactor`。


## 运行方式

1. 上传 `src/electronic_scale_and_weight_recording_system/electronic_scale_and_weight_recording_system.ino`。
2. 上电后先空载去皮。
3. 放置标准重量，使用校准加减按钮调整校准系数。
4. 放置物体后观察重量显示是否稳定。
5. 按记录按钮验证串口日志输出和记录次数累加。


## 控制逻辑说明

- 称重值通过 HX711 多次平均读取，减少抖动。
- 去皮键直接重置零点，便于容器称重场景。
- 校准按键动态修改 `calibrationFactor`，无需重新烧录代码。
- 记录键会把当前重量通过串口输出，方便外部保存数据。
- LCD 始终显示当前重量和关键参数，便于现场标定与记录。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 重量采集 | 放置砝码后显示数值变化明显 |
| 去皮功能 | 去皮后空盘重量回到接近 0 |
| 校准调节 | 修改校准系数后重量显示同步变化 |
| 记录输出 | 按记录按钮后串口输出称重值 |


## 可扩展方向

- 增加 EEPROM 保存校准系数。
- 增加 SD 卡本地称重记录。
- 增加单位切换和累计计量功能。
- 接入上位机形成物料称重终端。
