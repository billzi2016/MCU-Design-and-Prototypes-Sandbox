# 电子密码锁

## 项目简介

本项目实现一个基于 Arduino Uno 的电子密码锁系统，使用 4x4 矩阵键盘输入密码，LCD1602 I2C 显示操作状态，舵机模拟门锁开合，蜂鸣器用于按键反馈和错误报警。

系统支持密码验证、开锁控制、错误次数限制、临时锁定和密码修改。Arduino Uno 可以满足键盘扫描、LCD 显示、舵机控制和蜂鸣器输出需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 4x4 矩阵键盘 | 1 | 密码输入和功能按键 |
| LCD1602 I2C | 1 | 显示输入状态和系统提示 |
| SG90 舵机 | 1 | 模拟门锁开合 |
| 有源蜂鸣器 | 1 | 按键反馈和错误提示 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 支持 4 位数字密码输入。
- LCD 显示输入状态，密码以 `*` 隐藏。
- 密码正确后舵机转动开锁，延时后自动关锁。
- 密码错误后蜂鸣器提示。
- 连续错误 3 次后进入短暂锁定状态。
- `#` 键确认输入。
- `*` 键清空当前输入。
- `A` 键进入密码修改流程。
- 密码修改需要先输入旧密码，再输入新密码。

## 接线说明

### LCD1602 I2C 接线

| LCD1602 I2C | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| SDA | A4 |
| SCL | A5 |

### 4x4 矩阵键盘接线

矩阵键盘通常有 8 个引脚，分别对应 4 行和 4 列。不同键盘模块排针顺序可能不同，接线前应核对模块丝印或用万用表确认行列关系。

| 键盘引脚 | Arduino Uno |
| --- | --- |
| R1 | D2 |
| R2 | D3 |
| R3 | D4 |
| R4 | D5 |
| C1 | D6 |
| C2 | D7 |
| C3 | D8 |
| C4 | D9 |

### 舵机接线

| SG90 舵机 | Arduino Uno |
| --- | --- |
| 信号线 | D10 |
| VCC | 5V |
| GND | GND |

舵机启动电流较大，如果舵机抖动或 Arduino 重启，应使用独立 5V 电源给舵机供电，并确保外部电源 GND 与 Arduino GND 共地。

### 蜂鸣器接线

| 蜂鸣器 | Arduino Uno |
| --- | --- |
| 正极 | D11 |
| 负极 | GND |


## 运行方式

1. 使用 Arduino IDE 打开 `src/electronic_password_lock/electronic_password_lock.ino`。
2. 安装 `Keypad` 库。
3. 安装 `LiquidCrystal_I2C` 库。
4. 安装 `Servo` 库，Arduino IDE 通常已内置。
5. 开发板选择 `Arduino Uno`。
6. 按照接线说明完成连接。
7. 编译并上传程序。


## 状态设计

| 状态 | 说明 |
| --- | --- |
| 正常输入 | 输入密码并等待确认 |
| 开锁 | 密码正确，舵机转到开锁角度 |
| 锁定 | 连续错误达到限制，短时间拒绝输入 |
| 修改密码验证 | 修改密码前先验证旧密码 |
| 修改密码输入 | 输入新密码并保存 |


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 密码输入 | LCD 显示对应数量的 `*` |
| 正确密码 | 舵机转到开锁角度，随后自动关锁 |
| 错误密码 | 蜂鸣器提示，错误次数增加 |
| 连续错误 | 连续错误 3 次后进入锁定状态 |
| 清空输入 | `*` 键可以清除当前输入 |
| 修改密码 | 输入旧密码后可以保存新密码 |
| 新密码生效 | 使用新密码可以正常开锁 |


## 可扩展方向

- 使用 EEPROM 保存密码，断电后不丢失。
- 增加 RFID 解锁。
- 增加指纹模块。
- 增加蓝牙临时密码。
- 使用继电器或电磁锁替代舵机。
- 增加开锁记录和异常报警。
