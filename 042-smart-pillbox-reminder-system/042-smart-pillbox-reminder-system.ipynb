# 智能药盒提醒系统

## 项目简介

本项目实现一个基于 Arduino、RTC、舵机、蜂鸣器和 OLED 的智能药盒提醒系统。系统按照预设时间到点提醒用户服药，并打开对应药仓；若用户在限定时间内确认服药，则记为完成，否则记录为 missed dose，用于后续看护统计。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| DS3231 RTC | 1 | 实时时钟 |
| 舵机 | 2 | 控制两个药仓开合 |
| 蜂鸣器 | 1 | 提醒音 |
| OLED | 1 | 时间与状态显示 |
| 按键 | 2 | 服药确认和页面切换 |


## 核心功能

- 根据 RTC 当前时间触发定时提醒。
- 打开对应药仓并启动蜂鸣提示。
- 用户确认后记录当次服药完成。
- 超时未确认则记为 missed dose。
- OLED 显示时间、计划、完成次数和漏服次数。


## 引脚分配

| 模块 | 引脚 |
| --- | --- |
| 蜂鸣器 | D3 |
| 药仓舵机 1 / 2 | D5、D6 |
| 确认按键 | D8 |
| 页面切换按键 | D9 |
| DS3231 SDA / SCL | I2C 总线 |
| OLED SDA / SCL | I2C 总线 |


## 接线说明

- DS3231 和 OLED 均通过 I2C 接入主控板。
- 两路舵机分别控制两个药仓盖，建议独立 5V 供电并与主控共地。
- 确认按键和页面切换按键接 D8、D9，使用内部上拉。
- 药仓开合角度需要根据实际机械结构重新标定。


## 运行方式

1. 上传 `src/smart_pillbox_reminder_system/smart_pillbox_reminder_system.ino`。
2. 先校准 RTC 当前时间，再检查 OLED 显示是否正常。
3. 调整两组用药计划时间，便于短周期测试提醒功能。
4. 观察到点后药仓是否打开、蜂鸣器是否响起。
5. 分别测试确认服药和超时漏服两种场景。


## 控制逻辑说明

- 系统维护 `IDLE / REMINDING / MISSED` 三种提醒状态。
- 每条提醒计划每天只触发一次，到第二天自动复位。
- 提醒开始后舵机会打开指定药仓，并启动蜂鸣器提示。
- 用户在窗口内按确认键则记为已服药，否则超时记为 missed dose。
- OLED 轮播显示当前时间、今日计划完成情况和累计统计信息。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 定时提醒 | 到设定时间会触发提醒 |
| 药仓动作 | 对应药仓会打开 |
| 服药确认 | 按确认键后计入已服药次数 |
| 漏服记录 | 超时未确认后 missed dose 次数增加 |


## 可扩展方向

- 增加短信或蓝牙提醒。
- 增加更多药仓和可编辑计划。
- 增加看护端同步与历史记录导出。
- 增加重量检测确认是否真正取药。
