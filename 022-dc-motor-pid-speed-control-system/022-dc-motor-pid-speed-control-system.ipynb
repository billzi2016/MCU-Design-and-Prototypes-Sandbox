# 直流电机 PID 调速系统

## 项目简介

本项目实现一个基于 Arduino Uno 的直流电机 PID 调速系统，使用编码器脉冲估算电机实时转速，使用 PID 算法调节 PWM 输出，使电机尽量稳定在目标转速附近。系统通过串口输出目标速度、实际速度和 PWM，便于参数整定和调试。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| 直流电机 | 1 | 被控对象 |
| 编码器 | 1 | 速度反馈 |
| 电机驱动 | 1 | PWM 输出驱动 |
| 按键 | 2 | 切换目标和使能 |


## 核心功能

- 编码器测速。
- PWM 调速。
- PID 闭环控制。
- 串口调试输出。
- 目标转速切换。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| 电机 PWM | D5 |
| 电机方向 | D6、D7 |
| 编码器脉冲输入 | D2 |
| 目标切换按键 | D8 |
| 使能按键 | D9 |


## 接线说明

编码器信号接 D2，电机驱动 PWM 与方向引脚接 D5、D6、D7；两个按键使用内部上拉方式接 D8、D9。电机电源建议独立供电并与 Arduino 共地。


## 运行方式

1. 打开 `src/dc_motor_pid_speed_control_system/dc_motor_pid_speed_control_system.ino`。
2. 安装 `PID_v1` 库。
3. 接好编码器、驱动和电机。
4. 编译上传后观察串口中的转速与 PWM 输出。


## 控制逻辑说明

- 固定时间窗统计编码器脉冲增量，换算实时转速。
- PID 把目标转速与实际转速的误差转换成 PWM 输出。
- 电机失能时强制输出 0，便于做安全测试。
- 串口持续打印目标、测量值和输出值，便于在线整定参数。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 编码器测速 | 串口能输出稳定 RPM |
| 目标切换 | 按键切换后目标速度改变 |
| PID 调速 | 实际速度能跟随目标变化 |
| 使能控制 | 禁用后 PWM 归零 |
