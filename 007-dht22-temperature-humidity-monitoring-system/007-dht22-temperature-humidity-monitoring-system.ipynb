# DHT22 温湿度监测系统

## 项目简介

本项目实现一个基于 Arduino Uno 的 DHT22 温湿度监测系统，使用 DHT22 采集环境温度和湿度，使用 LCD1602 I2C 实时显示监测值和报警状态，使用蜂鸣器在超限时发出提示。

系统不仅显示实时数据，还加入了上下限阈值判断和回差处理，避免温湿度在临界点附近波动时报警频繁抖动。Arduino Uno 可以满足单总线传感器采样、LCD 显示和蜂鸣报警需求，因此选择 Uno 作为最小实现平台。


## 主要器件

| 器件 | 数量 | 说明 |
| --- | --- | --- |
| Arduino Uno | 1 | 主控板 |
| DHT22 | 1 | 温湿度采集 |
| LCD1602 I2C | 1 | 实时显示温湿度与状态 |
| 有源蜂鸣器 | 1 | 超限报警 |
| 10k 电阻 | 1 | DHT22 数据线上拉电阻 |
| 面包板和杜邦线 | 若干 | 原型连接 |


## 核心功能

- 实时采集环境温度。
- 实时采集环境湿度。
- 支持温度过高、过低报警。
- 支持湿度过高、过低报警。
- 支持多项超限同时报警。
- LCD1602 实时显示测量值与系统状态。
- 加入阈值回差，减少临界点来回跳变造成的误报警。
- DHT22 读数异常时显示传感器故障信息。


## 引脚分配

| 功能 | Arduino Uno 引脚 |
| --- | --- |
| DHT22 数据线 | D2 |
| 蜂鸣器 | D8 |
| LCD SDA | A4 |
| LCD SCL | A5 |


## 接线说明

### DHT22 接线

| DHT22 引脚 | Arduino Uno |
| --- | --- |
| VCC | 5V |
| DATA | D2 |
| GND | GND |

DHT22 的数据线与 VCC 之间建议接一个 10k 上拉电阻，提高通信稳定性。

### LCD1602 I2C 接线

| LCD1602 I2C | Arduino Uno |
| --- | --- |
| VCC | 5V |
| GND | GND |
| SDA | A4 |
| SCL | A5 |

### 蜂鸣器接线

| 蜂鸣器 | Arduino Uno |
| --- | --- |
| 正极 | D8 |
| 负极 | GND |


## 运行方式

1. 使用 Arduino IDE 打开 `src/dht22_temperature_humidity_monitoring_system/dht22_temperature_humidity_monitoring_system.ino`。
2. 安装 `DHT sensor library` 库。
3. 安装 `LiquidCrystal_I2C` 库。
4. 开发板选择 `Arduino Uno`。
5. 按照接线说明完成连接。
6. 编译并上传程序。


## 控制逻辑说明

- 每 2 秒读取一次 DHT22 数据。
- 对温度高、温度低、湿度高、湿度低分别独立判断。
- 每个报警阈值都设置回差，避免测量值刚好在临界点附近反复触发和解除报警。
- 蜂鸣器在报警状态下间歇鸣叫，正常状态下保持关闭。
- LCD 页面轮流显示实时测量值和当前阈值设置。
- 如果连续读取失败，系统进入传感器故障提示状态。


## 验证标准

| 测试项 | 通过标准 |
| --- | --- |
| 温度显示 | LCD 正确显示当前温度 |
| 湿度显示 | LCD 正确显示当前湿度 |
| 高温报警 | 温度高于阈值后蜂鸣器报警，LCD 显示高温状态 |
| 低温报警 | 温度低于阈值后蜂鸣器报警，LCD 显示低温状态 |
| 高湿报警 | 湿度高于阈值后蜂鸣器报警，LCD 显示高湿状态 |
| 低湿报警 | 湿度低于阈值后蜂鸣器报警，LCD 显示低湿状态 |
| 多项报警 | 同时超出多个阈值时显示多项报警状态 |
| 传感器异常 | 拔掉数据线或读数失败时，LCD 显示故障提示 |
